## 1. create a function to print first 10 natural numbers



```python
def func():
  print("first 10 natural numbers are: ",end=" ")
  for i in range(1,11):
    print(i,end=" ")

func()
```

## 2. create a function to calculate sum of first N natural numbers

```python
def SUM(n):
  print(f"first {n} natural numbers are: ",end=" ")
  sum = 0
  for i in range(1,n+1):
    sum += i
  print(sum,end=" ")
n = int(input("enter the value of n: "))
SUM(n)
```

## 3. create a function to reverse a number

```python
def Reverse(n):
  r = 0
  while n>0:
    s = n%10
    r = r*10 + s
    n = n//10
  print(r)
n = int(input("enter the number: "))
Reverse(n)
```

## 4. create a function to count digits in a number

```python
def Count(n):
  n = abs(n)
  count = 0
  if n == 0:
    return 1
  while n>0:
    count = count + 1
    n = n//10
  print(count)
n = int(input("enter the number: "))
Count(n)
```

## 5. create a function to check palindrome

```python
def palindrome(n):
  r = 0
  a = n
  while n>0:
    s = n%10
    r = r*10 + s
    n = n//10
  if r == a:
    return "palindrome"
  else:
    return "not palindrome"
n = int(input("enter the number: "))
palindrome(n)
```

## 6. create a function to generate fibonacci series

```python
def fibo(n):
  if n <= 0:
    return "invalid input"
  if n == 1:
    return 0
  s = [0,1]
  print(f"fibonacci series upto {n}:",0 ,1,end=' ')
  for i in range(2,n):
    s.append(s[i-1] + s[i-2])
    print(s[i],end=" ")
n = int(input("enter the number: "))
fibo(n)
```

## 11. Create a student class with name and marks

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    def display(self):
        print(f"Student name: {self.name}")
        print(f"Student age: {self.age}")
        print(f"Student grade: {self.grade}")

s1 = Student("Vandana",20,"A")
s2 = Student("Arnika",20,"A")
s1.display()
print("------------------------------")
s2.display()
```
