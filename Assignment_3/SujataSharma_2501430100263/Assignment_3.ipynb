```python
#2 create a function to calculate sum of first N natural numbers
def sum_numbers(n):
  sum=0
  for i in range(1,n+1):
   sum+=i
  print("sum of n natural numbers is:",sum)
n= int(input("enter the value of n upto which sum is required:"))
sum_numbers(n)
```

```python
#3 create a function to reverse a number
def reverse_number(n):
 rev=0
 while n >0:
  digit= n % 10
  rev= rev*10 + digit
  n= n // 10
 print("reverse of the number is:",rev)
n= int(input("enter a number:"))
reverse_number(n)
```

```python
#4 create a function to count digits in a number
def count_digits(n):
 a= str(n)
 b= list(a)
 length= len(b)
 print("number of digits in the number is:",length)
n= int(input("enter any number:"))
count_digits(n)
```

```python
#5 create a function to check palindrome number
def pal(n):
 rev=0
 n2=n
 while n2>0:
   n1= n2%10
   rev= rev*10+n1
   n2= n2//10
 if rev==n:
  print(n,"is a palindrome")
 else:
  print(n,"is not a palindrome")
n= int(input("enter a number:"))
pal(n)
```

```python
#6 create a function to generate fibonnaci series
def Fib(n):
  a,b=0,1
  for i in range(n):
    print(a,end=" ")
    a,b=b,a + b

n = int(input("enter the number of terms:"))
Fib(n)
```

```python
#7 calculator using functions that contains the following features;
  # - user selects operations
  # - program performs calculation
  # - Display result
a= int(input("enter first number:"))
b= int(input("enter second number:"))
print("Choose Add=1,Sub=2,Multiplication=3, Division=4")
operation= int(input("choose operation:"))
if operation == 1:
  print("result:",a+b)
elif operation ==2:
  print("result:", a-b)
elif operation == 3:
  print("result:", a*b)
elif operation==4:
  print("result:",a/b)
else:
  print("Enter correct operation")
```

```python
#11 create a student class with name and marks
class Student:
  def __init__(self,name,marks):
    self.name=name
    self.marks=marks
  def display(self):
    print("Name:",self.name,"\nMarks:", self.marks)
name= input("Enter name of the student:")
marks= int(input("Enter marks obtained by the student: "))
details= Student(name, marks)
details.display()
```
