1. Function to Print First 10 Natural Numbers.

```python
def print_numbers():
    for i in range(1, 11):
        print(i)

print_numbers()
```

2. Function to Calculate Sum of First N Natural Numbers.

```python
def sum_natural(n):
    total = 0

    for i in range(1, n + 1):
        total = total + i

    return total


n = int(input("Enter a number: "))
print("Sum =", sum_natural(n))
```

3. Function to Reverse a Number.

```python
def reverse_number(num):
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10

    return reverse


number = int(input("Enter a number: "))
print("Reversed Number =", reverse_number(number))
```

4. Function to Count Digits in a Number.

```python
def count_digits(num):
    count = 0

    while num > 0:
        num = num // 10
        count += 1

    return count


number = int(input("Enter a number: "))
print("Total Digits =", count_digits(number))
```

5. Function to Check Palindrome Number.

```python
def palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10

    if original == reverse:
        return True
    else:
        return False


number = int(input("Enter a number: "))

if palindrome(number):
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

6. Function to Generate Fibonacci Series.

```python
def fibonacci(n):
    a = 0
    b = 1

    print("Fibonacci Series:")

    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c


num = int(input("Enter number of terms: "))
fibonacci(num)
```

7. Calculator Using Functions.

```python
def add(a, b):
    return a + b


def subtract(a, b):
    return a - b


def multiply(a, b):
    return a * b


def divide(a, b):
    return a / b


print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("Enter your choice: "))

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == 1:
    print("Result =", add(num1, num2))

elif choice == 2:
    print("Result =", subtract(num1, num2))

elif choice == 3:
    print("Result =", multiply(num1, num2))

elif choice == 4:
    print("Result =", divide(num1, num2))

else:
    print("Invalid Choice")
```

8. Create a Text File and Store Student Details.

```python
file = open("student.txt", "w")

name = input("Enter student name: ")
roll = input("Enter roll number: ")
marks = input("Enter marks: ")

file.write("Name: " + name + "\n")
file.write("Roll Number: " + roll + "\n")
file.write("Marks: " + marks)

file.close()

print("Student details saved successfully.")
```

11. Create a Student Class with Name and Marks.

```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Student Name:", self.name)
        print("Student Marks:", self.marks)


name = input("Enter student name: ")
marks = int(input("Enter marks: "))

s1 = Student(name, marks)

s1.display()
```
