```python
def sum(n):#Sum of first n natural numbers
  s=0
  for i in range(1,n+1):
    s+=i
  return s
result = sum(6)
print(result)
```

```python
def reverse(n):#Reversing a nummber
  r = 0
  while n>0:
    digit=n%10
    r = r*10+digit
    n = n//10
  print(r)
reverse(1205)

```

```python
def palindrome(n):#Checking a palindrome number
  n = str(n)
  left = 0
  right = len(n)-1
  for i in range(len(n)):
    while(left<=right):
     if n[left]!=n[right]:
      return False
     else:
      left+=1
      right-=1
  return True
result = palindrome(5203)
if result==True:
   print("Yes the number is palindrome")
else:
  print("No it is not")
```
