```python
#Q2
def sum_naturalno(n):
  sum = 0
  for i in range(1,n+1):
    sum += i
  return sum

n = int(input("Enter value of n to calculate sum : "))
result = sum_naturalno(n)
print("The sum of first ", n ,"natural numbers is : ",result)
```

```python
#Q3
def reverse_number(num):
    rev = 0
    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10

    return rev

n = int(input("Enter a number: "))
result = reverse_number(n)

print("Reversed number is:", result)
```

```python
#Q4
def count_digit(n):
  if n == 0:
    count = 1

  else :
    count = 0
    while n>0:
      n = n//10
      count += 1
  return count

n = int(input("Enter a number : "))
result = count_digit(n)
print("The number of digits in the given number are : ",result)
```

```python
#Q5
def palindrome(n):
  original = n
  reverse = 0
  while n > 0:
    digit = n % 10
    reverse = reverse*10 + digit
    n = n//10

  if original == reverse:
    print("Palindrome number.")
  else:
    print("Not a Palindrome number.")

n = int(input("Enter a number : "))
palindrome(n)
```

```python
#Q6
def fibonacci(n):
    a, b = 0, 1

    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
n = int(input("Enter number of terms in series : "))
fibonacci(n)
```

```python
#Q7
def add(a,b):
  return a + b
def sub(a,b):
  return a - b
def mul(a,b):
  return a * b
def div(a,b):
  if b == 0:
    return "Division by zero is not allowed"
  return a/b

a = int(input("Enter first number : "))
b = int(input("Enter second number : "))
choice = input("Enter the operation to be performed from (+,-,*,/) : ")
if choice == "+":
  print("Result = ",add(a,b))
elif choice == "-":
  print("Result = ",sub(a,b))
elif choice == "*":
  print("Result = ",mul(a,b))
elif choice == "/":
  print("Result = ",div(a,b))
else:
  print("Invalid Choice!!")



```

```python
#Q8
import pandas as pd
def student(n):
  list1 = []
  list2 =[]
  list3 = []
  for i in range(n):
    name = input(f"Enter name of student {i+1} : " )
    age = int(input(f"Enter age of student {i+1} : " ))
    city = input(f"Enter city of student {i+1} : " )
    list1.append(name)
    list2.append(age)
    list3.append(city)
  data = {
      "Name" : list1,
      "Age" : list2,
      "City": list3
  }
  return data

n = int(input("Enter number of students : "))
result = student(n)
df = pd.DataFrame(result)
df.to_csv("students.txt", sep="\t", index=False)
print("Data saved to student.txt")
```
