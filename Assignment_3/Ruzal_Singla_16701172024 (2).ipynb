Sum of first N natural numbers

```python
def sum_natural(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

n = int(input("Enter N: "))
print(f"Sum = {sum_natural(n)}")
```

Reverse a number

```python
def reverse_number(n):
    rev = 0
    while n > 0:
        digit = n % 10
        rev   = rev * 10 + digit
        n     = n // 10
    return rev

n = int(input("Enter number: "))
print(f"Reversed = {reverse_number(n)}")
```

Count digits in a number

```python
def count_digits(n):
    count = 0
    while n > 0:
        n //= 10
        count += 1
    return count

n = int(input("Enter number: "))
print(f"Digits = {count_digits(n)}")
```

Check palindrome number

```python
def is_palindrome(n):
    original = n
    rev = 0
    while n > 0:
        digit = n % 10
        rev   = rev * 10 + digit
        n     = n // 10
    return original == rev

n = int(input("Enter number: "))
if is_palindrome(n):
    print(f"{n} is a Palindrome")
else:
    print(f"{n} is not a Palindrome")
```

Fibonacci series

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print()

n = int(input("How many terms? "))
fibonacci(n)
```


Calculator using functions

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b

def calculator():
    while True:
        print("\n===== CALCULATOR =====")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        print("5. Exit")

        choice = input("Select operation (1-5): ")

        if choice == "5":
            print("Goodbye!")
            break

        if choice in ["1", "2", "3", "4"]:
            a = float(input("Enter first number : "))
            b = float(input("Enter second number: "))

            if choice == "1":
                print(f"Result = {add(a, b)}")
            elif choice == "2":
                print(f"Result = {subtract(a, b)}")
            elif choice == "3":
                print(f"Result = {multiply(a, b)}")
            elif choice == "4":
                print(f"Result = {divide(a, b)}")
        else:
            print("Invalid choice. Try again.")

calculator()
```

Handle division by zero using exception handling

```python
def safe_divide(a, b):
    try:
        result = a / b
        print(f"Result = {result}")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
    finally:
        print("Division operation complete.")

a = float(input("Enter numerator  : "))
b = float(input("Enter denominator: "))
safe_divide(a, b)
```

Student class with name and marks

```python
class Student:

    def __init__(self, name, marks):
        self.name  = name
        self.marks = marks

    def get_percentage(self):
        return (self.marks / 100) * 100

    def get_grade(self):
        p = self.get_percentage()
        if p >= 90:
            return "A+"
        elif p >= 75:
            return "A"
        elif p >= 60:
            return "B"
        elif p >= 40:
            return "C"
        else:
            return "F"

    def display(self):
        print(f"Name  : {self.name}")
        print(f"Marks : {self.marks}")
        print(f"Grade : {self.get_grade()}")

name  = input("Enter name  : ")
marks = float(input("Enter marks : "))

s1 = Student(name, marks)
s1.display()

```
