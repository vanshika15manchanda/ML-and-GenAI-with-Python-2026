```python
#1 sum of first 10 natural numbers
sum=0
for i in range(1,11):
  sum+=i
  i+=1
print(sum)
```

```python
#2 Find factorial of a number.
num=int(input("Enter a number:"))
fact=1
for i in range(1,num+1):
  fact*=i
  i+=1
print(fact)
```

```python
#3 Print fibonacci series
num1=0
num2=1
n=int(input("Enter the number of elemetns you want in the fibonacci series: "))
for i in range(n):
    print(num1, end=" ")
    num1, num2 = num2, num1+num2
```

```python
#4 find alrgest among 3 numbers
num1=int(input("Enter first number: "))
num2=int(input("Enter second number: "))
num3=int(input("Enter third number: "))
max=num1
if num2>num1 and num2>num3:
  max=num2
if num3>num1 and num3>num2:
  max=num3
print("The alrgest of the three given numbers is:",max)
```

```python
#5. Create student result system
name=input("Enter student name: ")
num=int(input("Enter the number of subejcts: "))
sum=0
for i in range(1,num+1):
  marks=int(input("Enter the marks of "+str(i)+" subject: "))
  sum+=marks

perc=(sum/(num*100))*100
print("Percentage of "+name+" is", perc)
if perc>=90:
  print("Grade A")
elif perc>=80 and perc<90:
  print("Grade B")
elif perc>=70 and perc<80:
  print("Grade C")
elif perc>=60 and perc<70:
  print("Grade D")
else:
  print("Grade F(FAIL)")
```
