```python
def fibonacci_series(n):
    a, b = 0, 1
    fib_list = []
    while len(fib_list) < n:
        fib_list.append(a)
        a, b = b, a + b
    return fib_list

# Generate the first 10 numbers in the Fibonacci series
n_terms = 10
series = fibonacci_series(n_terms)
print(f"Fibonacci series up to {n_terms} terms: {series}")
```

```python
Name=input("enter your name:")
Roll_no=int(input("enter your roll no:"))
marks_sub1=int(input("enter your marks sub1:"))
marks_sub2=int(input("enter your marks sub2:"))
marks_sub3=int(input("enter your marks sub3:"))

percentage=(marks_sub1+marks_sub2+marks_sub3)/300*100

def print_subject_grade(subject_name, marks):
  if marks>=90:
    print(f"grade A in {subject_name}")
  elif marks>=80:
    print(f"grade B in {subject_name}")
  elif marks>=70:
    print(f"grade C in {subject_name}")
  else:
    print(f"grade D in {subject_name}")

print_subject_grade("sub1", marks_sub1)
print_subject_grade("sub2", marks_sub2)
print_subject_grade("sub3", marks_sub3)

print(Name)
print(Roll_no)
print(percentage)
```
