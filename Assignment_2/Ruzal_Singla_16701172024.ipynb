
Sum of first 10 natural numbers

```python
total = 0

for i in range(1, 11):
    total += i

print("Sum = " , total)
```

Factorial of a number

```python
n = int(input("Enter n: "))

result = 1

for i in range(1, n + 1):
    result *= i

print("n! is ", result)
```


Fibonacci series

```python
n = int(input("How many terms? "))

a, b = 0, 1

print("Fibonacci:", end=" ")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

Largest among 3 numbers

```python
a = float(input("Enter number 1: "))
b = float(input("Enter number 2: "))
c = float(input("Enter number 3: "))

if a >= b and a >= c:
    print(f"Largest = {a}")
elif b >= a and b >= c:
    print(f"Largest = {b}")
else:
    print(f"Largest = {c}")
```

Student Result System

```python
name  = input("Enter name : ")
roll  = input("Enter roll : ")
sub1 = input("Enter sub1 : ")
sub2 = input("Enter sub2 : ")
sub3 = input("Enter sub3 : ")
sub4 = input("Enter sub4 : ")
sub5 = input("Enter sub5 : ")
total = int(sub1)+int(sub2)+int(sub3)+int(sub4)+int(sub5)
print("Name : ",name)
print("Roll : ",roll)
print("Total : ",total)
print("Percentage : ",(total/500)*100)

if (total/500)*100 > 90:
    print("Grade : A")
elif (total/500)*100 > 80:
    print("Grade : B")
elif (total/500)*100 > 70:
    print("Grade : C")
elif (total/500)*100 > 60:
    print("Grade : D")
else:
    print("Grade : F")
```
