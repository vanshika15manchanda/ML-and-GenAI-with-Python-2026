```python
#factorial
num=int(input("Enter a number: "))

#using for loop
fact=1
for i in range(num,0,-1):
  fact=fact*i
print("Factorial of",num,"using for loop: ",fact)


#using while loop
num_copy=num
fact_while=1
while num>0:
  fact_while*=num
  num=num-1
print("Factorial of",num_copy,"using while loop: ",fact_while)
```

```python
#fibonacci series
num=int(input("Enter a number: "))
a,b=0,1
if num==0:
  print(a)
elif num==b:
  print(a,b,end=" ")
else:
  print(a,b,end=" ")
  for i in range(2,num+1):

    z=a+b
    a,b=b,z
    print(z, end=" ")
```
