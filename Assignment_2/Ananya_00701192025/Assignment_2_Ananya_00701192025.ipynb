1. Find sum of first 10 natural numbers.

```python
total = 0
for i in range(1, 11):
    total = total + i

print("Sum of first 10 natural numbers =", total)
```

2. Find factorial of a number.

```python
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i

print("Factorial of", num, "=", factorial)
```

3. Find fibonacci series.

```python
n = int(input("Enter number of terms: "))
a = 0
b = 1
print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    
    c = a + b
    a = b
    b = c
```

4. Find largest among 3 numbers.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

if num1 >= num2 and num1 >= num3:
    largest = num1

elif num2 >= num1 and num2 >= num3:
    largest = num2

else:
    largest = num3
print("Largest number is =", largest)
```

5. Create Student Result System 
- Input student details 
- Input marks 
- Calculate percentage 
- Display grade

```python
# Student Result System

# Input student details
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")

# Input marks of 5 subjects
marks1 = float(input("Enter marks of Subject 1: "))
marks2 = float(input("Enter marks of Subject 2: "))
marks3 = float(input("Enter marks of Subject 3: "))
marks4 = float(input("Enter marks of Subject 4: "))
marks5 = float(input("Enter marks of Subject 5: "))

# Calculate total marks
total = marks1 + marks2 + marks3 + marks4 + marks5

# Calculate percentage
percentage = (total / 500) * 100

# Display grade
if percentage >= 90:
    grade = "A+"

elif percentage >= 80:
    grade = "A"

elif percentage >= 70:
    grade = "B"

elif percentage >= 60:
    grade = "C"

elif percentage >= 40:
    grade = "D"

else:
    grade = "Fail"

# Display result
print("\n----- Student Result -----")
print("Student Name :", name)
print("Roll Number  :", roll_no)

print("\nTotal Marks :", total)
print("Percentage  :", percentage, "%")
print("Grade       :", grade)
```
