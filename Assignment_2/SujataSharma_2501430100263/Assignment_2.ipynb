```python
#1 sum of first natural numbers
sum=0
for i in range(1,11):
 sum+=i #calculates the sum of natural numbers
print("sum of first 10 natural numbers is:",sum) #gives the output
```

```python
#2 find factorial of a number
n= int(input("enter a number:")) #takes a number as input
fact=1
for i in range(1,n+1):
 fact=fact*i #calculates the factorial of the given number
print("factorial of",n,"is:",fact)
```

```python
#3 print fibonacci series
f= [1,1]
n= int(input("enter number of terms:")) #it is the number of terms upto which we require
for i in range(n-2):
  f.append(f[-1]+f[-2])
print("the fibonacci series will be:",f)
```

```python
#4 find the largest among three numbers
p= int(input("enter first number:"))
q= int(input("enter second number:"))
r= int(input("enter third number:"))

if p>=q and p>=r: #compares p with other numbers
  print("largest number is",p)
elif q>=p and q>=r: #compares q with other numbers
  print("largest number is",q)
else:
  print("largest number is",r)
```

```python
#5 create a student result system and display the grades
name= input("enter student name:")
marks= input("Enter student marks of all subjects in dictionary:") #marks of the student is entered subject-wise
marks_dict = eval(marks)
marks_list=marks_dict.values()
total=0
for i in marks_list:
  total+=i
percent= total/len(marks_list)
print("percentage obtained by",name,"is:",percent)
if percent>=75 and percent<=100:
  print("grade obtained is A")
elif percent>=50 and percent<75:
  print("grade obtained is B")
elif percent>=25 and percent<50:
  print("grade obtained is C")
else:
  print("grade obtained is D")

```
