```python
 # Q-1 Find sum of first 10 natural numbers.

 sum = 0
for i in range(1, 11):
    sum += i
print("Sum =", sum)
```

```python
# Q-2 Find factorial of a number.

n = 5
fact = 1

for i in range(1, n+1):
    fact = fact * i

print("Factorial =", fact)
```

```python
# Q-3 print fibonacci series.

a = 0
b = 1

for i in range(10):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

```python
# Q-4 find largest among 3 numbers.

a = 10
b = 20
c = 15

if a >= b and a >= c:
    print(a)
elif b >= c:
    print(b)
else:
    print(c)
```

```python
# Q-5 create student result system.

name = input("Enter Name: ")

total = 0

for i in range(5):
    marks = int(input("Enter Marks: "))
    total += marks

percentage = total / 5

if percentage >= 90:
    grade = "A+"
elif percentage >= 75:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 40:
    grade = "C"
else:
    grade = "Fail"

print("Name:", name)
print("Percentage:", percentage)
print("Grade:", grade)
```
