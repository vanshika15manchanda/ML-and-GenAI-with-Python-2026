```python
#question no 3
num = int(input("enter your number: "))
nums = []
while num > 0:
  dig = num % 10
  nums.append(dig)
  num = num//10

ans = 0
for i in range(0,len(nums)):
  ans = ans*10 + nums[i]

print(ans)
```

```python
#question no 4
def func(n):
  dig = []
  while n > 0:
    d = n % 10
    dig.append(d)
    n = n//10
  reverse = dig[::-1]
  if dig == reverse:
    print("yes it is a palindrome")
  else:
    print("no it is not a palindrome")


num = int(input("enter your number: "))
func(num)
```

```python
#question no 5
n = int(input("enter the number of digits to be generated: "))
def fibonacci(n):
  a = 0
  b = 1
  for i in range(n):
    print(a)
    temp = a + b
    a = b
    b = temp

fibonacci(5)
```
