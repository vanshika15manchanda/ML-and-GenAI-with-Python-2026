# 1. Function to print first 10 natural numbers

```python
def print_natural_numbers():
    print("First 10 Natural Numbers:")
    for i in range(1, 11):
        print(i, end=" ")
    print()
```

# 2. Create a function to calculate sum of first N natural numbers.

```python
def sum_natural_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

sum_natural_numbers(12)
```

# 3. Create a function to reverse a number.

```python
def reverse_number(num):
    rev = 0
    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num //= 10
    return rev

reverse_number(7885654)
```

# 4. Create a function to count digits in a number.

```python
def count_digits(num):
    count = 0
    while num > 0:
        count += 1
        num //= 10
    return count

count_digits(4578378792)
```

# 5. Create a function to check palindrome number.

```python
def is_palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num //= 10

    if original == reverse:
        return True
    else:
        return False


# Main Program
number = int(input("Enter a number: "))

if is_palindrome(number):
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

# 6. Create a function to generate Fibonacci series.

```python
def fibonacci(n):
    a, b = 0, 1
    print("Fibonacci Series:")
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print()

fibonacci(17)
```

7. Calculator Using Functions that contains the following features;
- User selects operation
- Program performs calculation
- Display result

```python
# Calculator Using Functions

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero"


# Menu
print("Calculator")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("Enter your choice (1-4): "))

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == 1:
    result = add(num1, num2)
    print("Result =", result)

elif choice == 2:
    result = subtract(num1, num2)
    print("Result =", result)

elif choice == 3:
    result = multiply(num1, num2)
    print("Result =", result)

elif choice == 4:
    result = divide(num1, num2)
    print("Result =", result)

else:
    print("Invalid Choice")
```

# 8. Create a Text File and Store Student Details

```python
# Creating and writing student details to a file

file = open("student.txt", "w")

name = input("Enter student name: ")
roll_no = input("Enter roll number: ")
marks = input("Enter marks: ")

file.write("Name: " + name + "\n")
file.write("Roll No: " + roll_no + "\n")
file.write("Marks: " + marks + "\n")

file.close()

print("Student details saved successfully.")
```

# 10. Handle division by zero using exception handling.

```python
try:                                             #Code that may cause an error.
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))

    result = num1 / num2

    print("Result =", result)

except ZeroDivisionError:                        #Runs if denominator is 0.
    print("Error: Division by zero is not allowed.")

except ValueError:                               #Runs if user enters text instead of numbers.
    print("Error: Please enter valid numbers.")
```

# 11. Create a Student class with name and marks.

```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)


# Creating object
name = input("Enter student name: ")
marks = int(input("Enter marks: "))

s1 = Student(name, marks)

s1.display()
```
