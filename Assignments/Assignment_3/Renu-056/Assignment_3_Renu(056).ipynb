Q-2 Create a function to calculate sum of first N  natural number.

```python
def sum_calculator(n):
  sum=0
  for i in range(1,n+1):

    sum+=i
  print(sum)

sum_calculator(10)
```

Q-3 Create a function to reverse a number.

```python
def reverse_number(n):
  print(str(n)[::-1])

reverse_number(745689)
```

Q-5 Create a function to check palindrome number.

```python
def palindrome(n):
  if str(n)==str(n)[::-1]:
    print("The given number is a palinfrome.")
  else:
    print("The given number is not a palindrome.")

palindrome(7888778887)
```

Q-6 Create a function to generate Fibonacci Series.

```python
def fibonacci(n):
  a=0
  b=1

  for i in range (n):
    print(a,end=" ")

    c=a+b
    a=b
    b=c
fibonacci(10)

```

Q-7 Calculator using functions that contains the following features:   
-- User selects operations
--program performs calculation:
--display result

```python
def calculator(a,b):
  operation=input("Enter what you want to perform:")
  if operation=="add":
    print(a+b)
  elif operation=="subtract":
    print(a-b)
  elif operation=="multiplication":
    print(a*b)
  elif operation=="divide":
    if b==0:
      print("Division can't be performed.")
    else:
      print(a/b)

calculator(42,7)
```
