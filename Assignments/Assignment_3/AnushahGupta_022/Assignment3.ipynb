```python
#function to print the first 10 natural numbers
def print_natural_numbers():
    print("First 10 Natural Numbers:")
    for i in range(1, 11):
        print(i)
print_natural_numbers()
```

```python
#function to calculate the sum of n natural numbers
def sum_natural_numbers(n):
    sum=0
    for i in range(1,n+1):
      sum=sum+i
    return sum


print(sum_natural_numbers(3))
```

```python
#function to reverse a number
def reverse_number(num):
    rev = 0
    while num > 0:
        digit = num % 10
        rev = rev*10 + digit
        num //= 10
    return rev

print(reverse_number(621))
```

```python
#function to count digits in a number

def count_digits(num):
    count = 0
    while num > 0:
        count =count+1
        num //= 10
    return count

print(count_digits(126))
```

```python
#function to check palindrome number
def palindrome(num):
    original = num
    reverse = 0

#reverse the number
    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10

#check if they are the same
    if original == reverse:
        print("it is a palindrome number")
    else:
        print("it is not a Palindrome Number")

palindrome(121)
```

```python
#function to generate fibonacci series
def fibonacci(n):
    a=0
    b=1
    for _ in range(n):
        print(a)
        a,b = b,a + b
fibonacci(5)
```

```python
#creating a calculator using functions

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
        return a / b

def calculator():
    print("\nCalculator")
    print("1.add")
    print("2.subtract")
    print("3.multiply")
    print("4.divide")

    choice = int(input("enter the operation (1-4):"))
    num1 = float(input("enter first number:"))
    num2 = float(input("enter second number:"))

    if choice == 1:
        print("Result =",add(num1,num2))
    elif choice == 2:
        print("Result =",subtract(num1,num2))
    elif choice == 3:
        print("Result =",multiply(num1,num2))
    elif choice == 4:
        print("Result =",divide(num1,num2))
    else:
        print("Invalid Choice")

calculator()
```

```python
#create a text file to store student details

def write_student_file():
    with open("student.txt", "w") as file:
        file.write("Name:Anushah\n")
        file.write("Roll No:022\n")
        file.write("Marks:73")
    print("Student details saved in student.txt")

write_student_file()
```

```python
#reading from a file
def read_student_file():
    try:
        with open("student.txt", "r") as file:
            print("student setails:")
            print(file.read())
    except FileNotFoundError:
        print("File was not found.")

read_student_file()
```

```python
#creating class of student
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name :", self.name)
        print("Marks:", self.marks)

#creating a student object
student1 = Student("Anushah", 95)
student1.display()
```
