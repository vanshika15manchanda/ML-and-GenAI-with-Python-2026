```python
def print_first_10_natural_numbers():
    """Prints numbers from 1 to 10."""
    print("First 10 natural numbers:")
    for i in range(1, 11):
        print(i, end=" ")
    print()  # For a clean newline

# Call the function
print_first_10_natural_numbers()
```

```python
def calculate_sum_of_n_natural_numbers(n):
    """Calculates and returns the sum of natural numbers up to n."""
    total_sum = 0
    for i in range(1, n + 1):
        total_sum += i
    return total_sum

# Take input from user and call the function
num = int(input("Enter the value of N: "))
if num < 1:
    print("Please enter a natural number greater than or equal to 1.")
else:
    result = calculate_sum_of_n_natural_numbers(num)
    print(f"The sum of the first {num} natural numbers is: {result}")
```

```python
def reverse_number(number):
    """Reverses the given integer using loop mathematics."""
    reversed_num = 0
    # Store sign to handle negative numbers if needed
    sign = -1 if number < 0 else 1
    number = abs(number)

    while number > 0:
        remainder = number % 10
        reversed_num = (reversed_num * 10) + remainder
        number = number // 10

    return reversed_num * sign

# Test the function
input_num = int(input("Enter an integer to reverse: "))
print("Reversed number:", reverse_number(input_num))
```

```python
def count_digits(number):
    """Counts the total number of digits in an integer."""
    number = abs(number)
    # Special handle case for 0
    if number == 0:
        return 1

    count = 0
    while number > 0:
        count += 1
        number = number // 10
    return count

# Test the function
input_num = int(input("Enter a number to count its digits: "))
print("Total number of digits:", count_digits(input_num))
```

```python
def is_palindrome(number):
    """Checks if a number reads the same backward as forward."""
    # Reusing the mathematical logic from our reverse function
    if number < 0:
        return False  # Negative numbers are generally not palindromes due to '-'
    return number == reverse_number(number)

# Test the function
input_num = int(input("Enter a number to check for palindrome: "))
if is_palindrome(input_num):
    print(f"{input_num} is a Palindrome number.")
else:
    print(f"{input_num} is NOT a Palindrome number.")
```

```python
def generate_fibonacci(terms):
    """Generates and displays the Fibonacci sequence up to specified terms."""
    n1, n2 = 0, 1
    count = 0

    if terms <= 0:
        print("Please enter a positive integer.")
    elif terms == 1:
        print("Fibonacci series: [0]")
    else:
        print("Fibonacci series:")
        while count < terms:
            print(n1, end=" ")
            nth = n1 + n2
            n1 = n2
            n2 = nth
            count += 1
        print()

# Test the function
num_terms = int(input("Enter the number of Fibonacci terms to generate: "))
generate_fibonacci(num_terms)
```

```python
# --- Mathematical Core Operation Functions ---
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero is undefined."
    return x / y

# --- Driver Main Calculator Interface Function ---
def calculator():
    print("\n--- Functional Menu Calculator ---")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    # User selects operation
    choice = input("Select operation option (1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        # Program performs calculation & displays result
        if choice == '1':
            print(f"Result: {num1} + {num2} = {add(num1, num2)}")
        elif choice == '2':
            print(f"Result: {num1} - {num2} = {subtract(num1, num2)}")
        elif choice == '3':
            print(f"Result: {num1} * {num2} = {multiply(num1, num2)}")
        elif choice == '4':
            print(f"Result: {num1} / {num2} = {divide(num1, num2)}")
    else:
        print("Invalid Selection! Execution terminated.")

# Run the calculator function
calculator()
```
