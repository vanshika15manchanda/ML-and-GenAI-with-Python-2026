```python
#1
def firstSeries(n):
  print("First",n,"natural numbers:")
  for i in range(1,n+1):
    print(i, end=", ")

firstSeries(10)
```

```python
#2
def firstSum(n):
   s=0
   for i in range(1,n+1):
     s+=i
   print("Sum of first",n,"natural numbers:", s)
firstSum(12)
```

```python
#3
def reverseNum(n):
  print("Reverse of number:", str(n)[::-1])

reverseNum(679)
```

```python
#4
def no_of_digits(n):
  print("No. of digits:", len(str(n)))

no_of_digits(4510)
```

```python
#5
def checkPalindrome(n):
  if str(n)[::-1]==str(n):
    print(n,"is a palindrome.")
  else:
    print(n,"is not a palindrome.")

checkPalindrome(5665)
checkPalindrome(768)
```

```python
#7
def add(a, b):
  return a+b
def sub(a, b):
  return a-b
def mul(a, b):
  return a*b
def div(a, b):
  return a/b if b!=0 else "Error"

def Calculate():
    print("Select: 1.Add, 2.Sub, 3.Mul, 4.Div")
    ch = input("Enter choice (1-4): ")
    try:
        n1 = float(input("Enter num 1: "))
        n2 = float(input("Enter num 2: "))
        if ch == '1':
          print(f"Result:",add(n1, n2))
        elif ch == '2':
          print(f"Result:",sub(n1, n2))
        elif ch == '3':
          print(f"Result:",mul(n1, n2))
        elif ch == '4':
          print(f"Result:",div(n1, n2))
    except ValueError:
        print("Invalid input")

Calculate()
```

```python
#11
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Student:",self.name,"\nMarks:",self.marks)

student1 = Student("Alice", 92)
student1.display()
```
