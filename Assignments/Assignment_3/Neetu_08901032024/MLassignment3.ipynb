```python
#create a function to print first 10 natural numbers
def first10():
  for i in range(1,11):
    print(i)
first10()
```

```python
#create a function to calculate sum of first N natural numbers
def sum_N(n):
  sum=0
  for i in range(1,n+1):
    sum=sum+i
  print(sum)
sum_N(5)
```

```python
#create a function to reverse a number
def reverse(num):
  if num<0:
    print("enter valid number")
  rev=0
  while num>0:
    rev= rev*10+ num%10
    num=num//10
  print(rev)
reverse(852)
```

```python
#create a function to count digits in a number
def count_digits(num):
  count=0
  while num>0:
    count+=1
    num//=10
  print(count)
count_digits(7392)
```

```python
#create a function to check pallindrome number
def pallindrome(num):
  rev=0
  number=num
  while number>0:
    rev= rev*10+ number%10
    number=number//10
  if rev==num:
    print("pallindrome")
  else:
    print("not pallindrome")
pallindrome(12321)
```

```python
#create a function to generate fibonacci series
def fibonacci(n):
  last=0
  prev=1
  i=1
  while i<=n:
    print(last)
    curr= prev+last
    last=prev
    prev=curr
    i+=1
fibonacci(7)
```

```python
#calculator using functions that perform following features:user selects operation,program performs calculation,display result
# Function definitions

def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b == 0:
        return "Division by zero is not allowed"
    return a / b

print("Simple Calculator")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
choice = int(input("Enter your choice (1-4): "))
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
if choice == 1:
    result = add(num1, num2)
    print("Result =", result)
elif choice == 2:
    result = subtract(num1, num2)
    print("Result =", result)
elif choice == 3:
    result = multiply(num1, num2)
    print("Result =", result)
elif choice == 4:
    result = divide(num1, num2)
    print("Result =", result)
else:
    print("Invalid choice!")
```

```python
#create a student class with name and marks
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
# Creating an object
s1 = Student("Neetu", 95)
s2 = Student("Shivani",97)
# Displaying student details
s1.display()
s2.display()
```
