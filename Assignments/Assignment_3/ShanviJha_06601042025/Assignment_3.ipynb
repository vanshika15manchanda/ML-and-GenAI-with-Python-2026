```python
# 1. Create a function to print first 10 natural numbers.
def print_first_10_natural_numbers():
    #Printing natural numbers from 1 to 10 using a loop.
    print("First 10 natural numbers:")
    for i in range(1, 11):
        print(i, end=" ")
#for eg.
print_first_10_natural_numbers()
```

```python
# 2. Create a function to calculate sum of first N natural numbers.
def sum_of_n_natural_numbers(n):
    #Calculating and returning the sum of natural numbers up to N.
    # Formula used: total = n * (n + 1) // 2
    # Loop  used :
    total_sum = 0
    for i in range(1, n + 1):
        total_sum += i
    return total_sum
```

```python
# 3. Create a function to reverse a number.
def reverse_number(num):
    #Reversing the digits of a given integer
    reversed_num = 0
    # Handling negative numbers if passed
    is_negative = num < 0
    num = abs(num)

    while num > 0:
        remainder = num % 10
        reversed_num = (reversed_num * 10) + remainder
        num = num // 10
```

```python
#4. Create a function to count digits in a number.
def count_digits(num):
    num = abs(num)
    if num == 0:
        return 1

    count = 0
    while num > 0:
        count += 1
        num = num // 10
    return count
```

```python
# 6. Create a function to generate Fibonacci series.
def generate_fibonacci(terms):
    if terms <= 0:
        print("Please enter a positive integer.")
        return

    series = []
    a, b = 0, 1
    for _ in range(terms):
        series.append(a)
        a, b = b, a + b

    print("Fibonacci series ({terms} terms): {series}")
```

```python
#7.
# CALCULATOR FUNCTIONS

def add(x, y):
    #Performs addition
    return x + y

def subtract(x, y):
    #Performs subtraction
    return x - y

def multiply(x, y):
    #Performs multiplication
    return x * y

def divide(x, y):
    #Performs division and handles division by zero
    if y == 0:
        return "Error! Division by zero is not allowed."
    return x / y


# MAIN PROGRAM FLOW

def calculator():
    print("=== Python Text Calculator ===")
    print("Select an operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    #  User inputs operation
    choice = input("Enter choice (1/2/3/4): ")

    # Checking if the user entered a valid option
    if choice in ['1', '2', '3', '4']:
        try:
            # Prompt user for numerical inputs
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter numbers only.")
            return

        # 2. Program performs calculation based on the choice
        if choice == '1':
            result = add(num1, num2)
            operation = "+"
        elif choice == '2':
            result = subtract(num1, num2)
            operation = "-"
        elif choice == '3':
            result = multiply(num1, num2)
            operation = "*"
        elif choice == '4':
            result = divide(num1, num2)
            operation = "/"

        # 3. Display result
        print("\n--- Result ---")
        if isinstance(result, str):
            print(result)  # Prints the division-by-zero error message
        else:
            print(f"{num1} {operation} {num2} = {result}")

    else:
        print("Invalid choice! Please select a valid operation option (1-4).")

# Run the calculator program
calculator()
```
