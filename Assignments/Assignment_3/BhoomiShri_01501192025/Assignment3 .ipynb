```python
# 1.Create a function to print first 10 natural numbers.
def function():
  for x in range(10):
    print(x+1, end=" ")
function()
```

```python
# 2.Create a function to calculate sum of first N natural numbers.
def fn(N):
  s=0
  for x in range(N):
    s+=(x+1)
  return(s)
fn(10)
```

```python
# 3.Create a function to reverse a number.
def reverse(N):
  x=str(N)[::-1]
  return int(x)
reverse(4567)
```

```python
# 4.Create a function to count digits in a number.
def count(N):
  d=len(str(N))
  print("No. of digits ")
  return d
count(4567)
```

```python
# 5.Create a function to check palindrome number.
def is_palindrome(N):
  if str(N) == str(N)[::-1] :
    print("The number is palindrome.")
  else:
    print("The number is not palindrome")
is_palindrome(45654)
```

```python
# 6.Create a function to generate Fibonacci series.
def fib_series(N):
  a=0
  b=1
  print(a,b, end=" ")
  for x in range(N):
    c=a+b
    print(c, end=" ")
    a=b
    b=c
fib_series(7)
```

```python
# 7.Calculator Using Functions that contains the following features;
#   - User selects operation
#   - Program performs calculation
#   - Display result
def calculate(a,b,op):
  match op:
    case "+":
        result = a + b
    case "-":
        result = a - b
    case "*":
        result = a * b
    case "/":
        result = a / b if b != 0 else "Error: Division by zero"
    case _:
        result = "Invalid operator"

  print(f"{a} {op} {b} = {result}")
print("----------------------------------------------------------------------------------------------------------------")
print("CALCULATOR")
print(" ")
a=float(input('Enter the number1 '))
op=input("Select operator(+,-,*,/) ")
b=float(input('Enter the number2 '))
calculate(a,b,op)
```
