```python
#Q1 function to print first 10 natural nos
# 1 to 10
def print_naturals():
    for i in range (1, 11):
        print(i)
print_naturals()
```

```python
#Q2 sum of first n natural nos 

def sum_of_n(n):
    return n * (n + 1) // 2
print(sum_of_n(10))
```

```python
#Q3 func to reverse a no
def reverse_num(num):
    rev_no = 0
    while num > 0:
        digit = num % 10
        rev_no = rev_no * 10 + digit
        num //= 10
    return rev_no
print(reverse_num(1234))
```

```python
#Q4 func to count digit in a no
def count_digit(num):
    count = 0
    while num > 0:
        count += 1
        num //= 10
    return count
print(count_digit(12345678))
```

```python
#Q5 func to check palindrome no
def is_palindrome(num):
    return num == reverse_num(num)
print(is_palindrome(121))
```

```python
#Q6 func to generate fibonacci
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(15)
```

```python
#Q7 clac unsing func
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"

def calculator():
    print("1.Add  2.Subtract  3.Multiply  4.Divide")
    choice = int(input("Enter choice: "))
    
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    
    if choice == 1:
        print(add(a, b))
    elif choice == 2:
        print(subtract(a, b))
    elif choice == 3:
        print(multiply(a, b))
    elif choice == 4:
        print(divide(a, b))
    else:
        print("Invalid choice")
```

```python
#Q8 create a file and store student detail
def write_file():
    with open("students.txt", "w") as file:
        student_name = input("Enter name: ")
        marks = input("Enter marks: ")
        file.write(f"Name: {student_name}, Marks: {marks}")
```

```python
#Q9 read data from a file
def read_file():
    with open("students.txt", "r") as file:
        print(file.read())
```

```python
#Q10 try exception handling
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    print(a / b)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed")
```
