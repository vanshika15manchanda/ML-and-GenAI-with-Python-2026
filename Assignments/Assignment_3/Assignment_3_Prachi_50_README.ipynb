1. Create a function to print first 10 natural numbers.

2.Create a function to calculate sum of first N natural numbers.

```
def sum_nat_num(N):
  return N*(N+1)//2
print(sum_nat_num(10))

```

3.Create a function to reverse a number.

```
def reverse_num(n):
   return int(str(n)[::-1])
print(reverse_num(12345))
```

4.Create a function to count digits in a number.

```
def count_digits(n):
  return len(str(abs(n)))
print(count_digits(-1234))
```

5.Create a function to check palindrome number.

```
def palindrome(n):
  return str(n)==str(n)[::-1]
print(palindrome(121))
```

6.Create a function to generate fibonacci series.

```
def fibonacci(n):
  a,b=0,1
  for i in range(n):
    a,b=b,a+b
  print(a,end=" ")
print(fibonacci(10))
```

7.Calculator using functions that contains the following features;
-User selects operation
-Program performs calculation
-Display results


```
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b): return a / b if b != 0 else "Error!"

print("1. Add  2. Subtract  3. Multiply  4. Divide")
choice = int(input("Select operation: "))
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

if choice == 1: print(add(a, b))
elif choice == 2: print(subtract(a, b))
elif choice == 3: print(multiply(a, b))
elif choice == 4: print(divide(a, b))
```

9.Handle division by zero using exception handling.

```
try:
  a=int(input("Enter number: "))
  b=int(input("Enter divisor: "))
  print(a/b)
except ZeroDivisionError:
  print("Error: Division by zero")
except ValueError:
  print("Error: Invalid input")
```
