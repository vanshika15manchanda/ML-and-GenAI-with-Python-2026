```python
#Reverse a number
num=int(input("Enter 3 digit number:"))
def reverse(num):
 rev=0
 while num>0:
   digit=num%10
   rev=rev*10+digit
   num=num//10
 print(rev)
reverse(num)
```

```python
#count digits in a number
num=int(input("Enter number:"))
def count(num):
  count=0
  while num>0:
    digit=num%10
    count+=1
    num=num//10
  print(count)
count(num)
```

```python
#check palindrome number
num=int(input("Enter number:"))
def palindrome(num):
 original=num
 rev=0
 while num>0:
  digit=num%10
  rev=rev*10+digit
  num=num//10
 if original== rev:
   print("Palindrome number")
 else:
   print("Not a Palindrome number")
palindrome(num)
```

```python
#Fibonacci Series
num=int(input("Enter number of terms:"))
def fibonacci(num):
  a=0
  b=1
  for i in range(num):
   sum=a+b
   print(a)
   a=b
   b=sum
fibonacci(num)

```
