
## 2. Function to Calculate Sum of First N Natural Numbers

```python
# Formula: Sum = N * (N + 1) / 2
def sum_of_n_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total = total + i
    return total

n = int(input("Enter value of N: "))
result = sum_of_n_numbers(n)
print(f"Sum of first {n} natural numbers = {result}")
```


##  3. Function to Reverse a Number

```python
# Example: 1234 → 4321
def reverse_number(num):
    reversed_num = 0
    while num > 0:
        last_digit = num % 10          # Extract last digit
        reversed_num = reversed_num * 10 + last_digit  # Build reversed number
        num = num // 10                # Remove last digit
    return reversed_num
# Take input from user
num = int(input("Enter a number to reverse: "))
result = reverse_number(num)
print(f"Reversed number = {result}")
```


## 4. Function to Count Digits in a Number

```python
# Example: 12345 has 5 digits
def count_digits(num):
    count = 0
    while num > 0:
        num = num // 10   # Remove last digit each iteration
        count = count + 1 # Increment digit count
    return count

num = int(input("Enter a number: "))
digits = count_digits(num)
print(f"Number of digits = {digits}")
```


##  5. Function to Check Palindrome Number

```python
# A palindrome reads the same forwards and backwards
# Example: 121, 1331, 12321 are palindromes
def is_palindrome(num):
    original = num       # Save original number
    reversed_num = 0

    # Reverse the number
    while num > 0:
        last_digit   = num % 10
        reversed_num = reversed_num * 10 + last_digit
        num          = num // 10

    # If original == reversed, it is a palindrome
    if original == reversed_num:
        return True
    else:
        return False

num = int(input("Enter a number to check palindrome: "))

if is_palindrome(num):
    print(f"{num} is a Palindrome ")
else:
    print(f"{num} is NOT a Palindrome ")
```


## 6. Function to Generate Fibonacci Series

```python
# Series: 0, 1, 1, 2, 3, 5, 8, 13, ...
# Each term = sum of the two preceding terms

def fibonacci_series(n):
    a, b = 0, 1
    series = []
    for i in range(n):
        series.append(a)
        a, b = b, a + b
    return series

n = int(input("Enter number of terms for Fibonacci series: "))
series = fibonacci_series(n)
print(f"Fibonacci Series ({n} terms): {series}")
```


## 7. Calculator Using Functions

```python
# User selects an operation, program performs calculation, result is displayed

# Define one function per operation
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero!"
    return a / b

# Display menu for user to select operation
print("========== Calculator ==========")
print("  1. Addition       (+)")
print("  2. Subtraction    (-)")
print("  3. Multiplication (×)")
print("  4. Division       (÷)")
print("================================")

# User selects operation
choice = int(input("Select operation (1/2/3/4): "))

# Take two numbers as input
num1 = float(input("Enter first number : "))
num2 = float(input("Enter second number: "))

# Perform calculation using if-elif-else
if choice == 1:
    result = add(num1, num2)
    operator = "+"
elif choice == 2:
    result = subtract(num1, num2)
    operator = "-"
elif choice == 3:
    result = multiply(num1, num2)
    operator = "×"
elif choice == 4:
    result = divide(num1, num2)
    operator = "÷"
else:
    result = "Invalid choice!"
    operator = "?"

# Display result
print(f"\nResult: {num1} {operator} {num2} = {result}")
```


## 10. Handle Division by Zero using Exception Handling

```python
# Program to handle Division by Zero using try-except (Exception Handling)

try:
    num1 = float(input("Enter numerator   : "))
    num2 = float(input("Enter denominator : "))
    result = num1 / num2
    print(f"Result = {num1} / {num2} = {result}")

except ZeroDivisionError:
    # This block runs if denominator is 0
    print("Error: Division by zero is not allowed!")

except ValueError:
    # This block runs if user enters non-numeric input
    print(" Error: Please enter valid numbers only!")

finally:
    # This block always runs
    print("Program execution completed.")
```


##  11. Create a Student Class with Name and Marks

```python
# Demonstrates Object-Oriented Programming (OOP) in Python

class Student:
    # Constructor: initializes name and marks when object is created
    def __init__(self, name, marks):
        self.name  = name    # Student's name
        self.marks = marks   # Student's marks

    # Method to calculate percentage
    def get_percentage(self):
        total      = sum(self.marks)       # Sum of all marks
        max_marks  = len(self.marks) * 100 # Max possible marks
        percentage = (total / max_marks) * 100
        return percentage

    # Method to determine grade using if-elif-else
    def get_grade(self):
        percentage = self.get_percentage()
        if percentage >= 90:
            return "A+"
        elif percentage >= 80:
            return "A"
        elif percentage >= 70:
            return "B"
        elif percentage >= 60:
            return "C"
        elif percentage >= 40:
            return "D"
        else:
            return "F"

    # Method to display student report
    def display_report(self):
        print("==============================")
        print("       STUDENT REPORT         ")
        print("==============================")
        print(f"  Name       : {self.name}")
        print(f"  Marks      : {self.marks}")
        print(f"  Total      : {sum(self.marks)} / {len(self.marks) * 100}")
        print(f"  Percentage : {self.get_percentage():.2f}%")
        print(f"  Grade      : {self.get_grade()}")
        print("==============================")


# --- Take student input and create Student object ---
name        = input("Enter Student Name: ")
num_subjects = int(input("Enter number of subjects: "))

marks = []
for i in range(num_subjects):
    m = float(input(f"  Enter marks for Subject {i+1} (out of 100): "))
    marks.append(m)

# Create Student object
student1 = Student(name, marks)

# Display report using class method
student1.display_report()
```
