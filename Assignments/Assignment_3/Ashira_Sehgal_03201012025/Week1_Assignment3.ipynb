```python
#To reverse a number
def rev():
  n = int(input("Enter a number"))
  rev_num=0
  while(n>0):
    d=n%10
    rev_num=(rev_num*10)+d
    n=n//10
  print("The reversed number is:",rev_num)
rev()
```

```python
#count number of digits in a number
def count():
  num=input("Enter number:")
  print("Number of digits in number taken as string input are:",len(num))

count()
```

```python
#to check palindrome number
def palindrome():
  num=input("Enter number:")
  j=len(num)-1
  for i in range(len(num)//2):
    if(num[i]!=num[j]):
       print("Not palindrome")
       return
    else:
      j=j-1
  print("it is a palindrome")
palindrome()
```

```python
#fibonacci series
def fibonacci():
  n = int(input("Enter number of terms:"))
  a=0
  b=1
  print("The fibonacci series is:")
  for i in range(n):
    print(a)
    a,b=b,a+b

fibonacci()
```

```python
#calculator program
def calculate():
  a=int(input("Enter number:"))
  b=int(input("Enter number:"))
  o = input("Enter the operation you want to perform (+,-,/,*,%)")
  if(o=="+"):
    print("Output is:",a+b)
  elif(o=="-"):
     print("Output is:",a-b)
  elif(o=="*"):
     print("Output is:",a*b)
  elif(o=="/"):
     print("Output is:",a/b)
  elif(o=="%"):
     print("Output is:",a%b)

calculate()
```
