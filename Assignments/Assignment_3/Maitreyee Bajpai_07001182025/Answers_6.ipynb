```python
def generate_fibonacci(n):
    fib_series = []
    a, b = 0, 1
    for _ in range(n):
        fib_series.append(a)
        a, b = b, a + b
    return fib_series

# Example usage
terms = int(input("Enter the number of terms for Fibonacci series: "))
print("Fibonacci series:", generate_fibonacci(terms))
```
