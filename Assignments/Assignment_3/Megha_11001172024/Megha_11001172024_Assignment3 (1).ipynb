```python
# 1.Create a function to print the first 10 natural numbers.
#_function_
def natural_number(n):
  for i in range(1,n+1):
    print(i, end =" ")

#_main_
num = int(input("Enter the number:"))
natural_number(num)
```

```python
# 2.Create a function to calculate the sum of the first N natural numbers.
#_function_
def sum_natural(n):
   total = 0

   for i in range(1, n + 1):
        total += i

   return total

#_main_
n = int(input("Enter N: "))
print("Sum =", sum_natural(n))
```

```python
# 3.Create a function to reverse a number.
def reverse_number(num):
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10


    return rev

#_main_
num = int(input("Enter a number: "))
print("Reversed Number: ", reverse_number(num))
```

```python
# 4.Create a function to count the digits in a number.
#_function_
def count_digits(num):
    count = 0
    while num > 0:
        count += 1
        num = num // 10

    return count

#_main_
num = int(input("Enter a number: "))
print("Number of digits: ", count_digits(num))
```

```python
# 5.Create a function to check whether a number is a palindrome
#_function_
def check_palindrome(num):
    rev = 0
    temp = num

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 1
    if temp == rev:
        return True
    else:
        return False

#_main_
num = int(input("Enter a number: "))
if check_palindrome(num):
    print(num, "is a palindrome")
else:
    print(num, "is not a palindrome")
```

```python
# 6.Create a function to generate the Fibonacci Series
def fibonacci(n):
    a = 0
    b = 1

    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c
#_main_
n = int(input("Enter number of terms: "))
fibonacci(n)
```
