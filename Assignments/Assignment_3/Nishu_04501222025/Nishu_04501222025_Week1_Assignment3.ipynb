```python
#1.Function to print first 10 natural numbers
def natural_numbers():
    for i in range(1,11):
        print(i)
natural_numbers()
```

```python
#2.Function to calculate sum of first N natural numbers
def sum_natural_numbers(n):
  return n*(n+1) // 2
n = int(input("Enter a number: "))
print("Sum:",sum_natural_numbers(n))
```

```python
#3.Function to reverse a number
def reverse_number(num):
  rev = 0
  while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
  return rev
num = int(input("Enter a number: "))
print("Reversed number:",reverse_number(num))
```

```python
#4.Function to count digits in a number
def count_digits(num):
  count = 0
  while num > 0:
    num //= 10
    count += 1
  return count
num = int(input("Enter a number: "))
print("Number of digits:",count_digits(num))
```

```python
#5.Function to check palindrome number
def is_palindrome(num):
  original = num
  reverse = 0
  while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10
  return original == reverse
num = int(input("Enter a number: "))
if is_palindrome(num):
  print("Palindrome Number")
else:
  print("Not a Palindrome Number")
```

```python
#6.Function to generate Fibonacci Series
def fibonacci(n):
  a,b = 0,1
  for _ in range(n):
    print(a,end=" ")
    a,b = b,a+b
n = int(input("Enter the number of terms: "))
print("Fibonacci Series:")
fibonacci(n)
```

```python
#7.Calculator using Functions
def add(num1,num2):
  return num1 + num2
def subtract(num1,num2):
  return num1 - num2
def multiply(num1,num2):
  return num1 * num2
def divide(num1,num2):
  if num2 == 0:
    return "Error: Division by zero"
  else:
    return num1 / num2
print("Calculator Menu:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
choice = int(input("Enter your choice (1-4): "))
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
if choice == 1:
  print("Addition")
  print("Result:",add(num1,num2))
elif choice == 2:
  print("Subtraction")
  print("Result:",subtract(num1,num2))
elif choice == 3:
  print("Multiplication")
  print("Result:",multiply(num1,num2))
elif choice == 4:
  print("Division")
  print("Result:",divide(num1,num2))
else:
  print("Invalid Choice")
```

```python
#8.Create a text file and store student details
file = open("student_details.txt","w")
for i in range(5):
  print("Enter details for student",i+1)
  name = input("Enter student name: ")
  roll_no = input("Enter roll number: ")
  marks = input("Enter marks(out of 100): ")
  file.write("Student Name: " + name + "\n")
  file.write("Roll No: " + roll_no + "\n")
  file.write("Marks: " + marks + "\n")
  file.write("\n")
  print()
file.close()
print("Student Details Saved Successfully.")
```

```python
#10.Handle division by zero using exception handling
try:
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))
  result = num1 / num2
  print("Result:",result)
except ZeroDivisionError:
  print("Error: Division by zero is not allowed.")
except ValueError:
  print("Error: Invalid input. Please enter valid numbers.")
```
