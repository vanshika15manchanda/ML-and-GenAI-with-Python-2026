```python
#Q1 Create a function to print first 10 natural numbers
def Ten_nat():
  for i in range(1,11):
    print(i,end=' ')
Ten_nat()
```

```python
#Q2 Create a function to calculate sum of first N natural numbers
def Sum_nat(N):
  sum=0;
  for i in range(1,N+1):
    sum+=i
  return sum
x=int(input("Enter N: "))
print("Sum of",x,"natural no.s: ",Sum_nat(x))
```

```python
#Q3 Create a function to reverse a number.
def Reverse(n):
  rev_num=0
  while n>=1:
    rev_num=(rev_num*10)+(n%10)
    n=int(n/10)
  return rev_num

x=int(input("Enter a number: "))
print("Reverse number: ",Reverse(x))
```

```python
#Q4 Create a function to count digits in a number.
def CountDigit(n):
  count=0
  while n>=1:
    count+=1
    n=int(n/10)
  return count
x=int(input("Enter a number: "))
print("No of digits in num: ",CountDigit(x))
```

```python
#Q5 Create a function to check palindrome number.
def reverse(n):
  rev_num=0
  while n>=1:
    rev_num=(rev_num*10)+(n%10)
    n=int(n/10)
  return rev_num

def palindrome(m):
  if m==reverse(m):
    print(m,"is a pallindrome")
  else:
    print(m,"is NOT a pallindrome")

print("Test 1")
palindrome(1010)
print("\nTest 2")
palindrome(123321)
print("\nTest 3")
palindrome(12311)
```

```python
#Q6 Create a function to generate Fibonacci series.
def fibonacci(n):
  a=0
  b=1
  print(a,b,end=' ')
  for i in range(0,n-2):
    c=b
    b=a+b
    a=c
    print(b,end=' ')

x=int(input("Enter no. of terms(>1): "))
fibonacci(x)
```

```python
#Q7 Calculator Using Functions.

#Defining functions
def add(a,b):
  print(a+b)
def minus(a,b):
  print(a-b)
def multiply(a,b):
  print(a*b)
def divide(a,b):
  if b==0:
    print("Cannot divide by Zero!")
  else:
    print(a/b)

#Entering inputs
x=float(input("Enter X: "))
y=float(input("Enter Y: "))
oper=input("Enter operation(+,-,*,/): ")

if oper=='+':
  add(x,y)
elif oper=='-':
  minus(x,y)
elif oper=='*':
  multiply(x,y)
elif oper=='/':
  divide(x,y)
else:
  print("Wrong operation!")
```

```python
#Q8 Create a text file and store student details
def write_students_to_file(filename):
    student_details = [
        "ID: 1  Name: Aman    Grade: A",
        "ID: 2  Name: Arav    Grade: B",
        "ID: 3  Name: Ayush   Grade: A+"
    ]
    with open(filename, "w") as file:  #Creates file if it doesn't exist
        for student in student_details:
            file.write(student + "\n")

    print(f"Student details successfully saved to {filename}")

write_students_to_file("students.txt")
```

```python
#Q9 Read data from a file
def read_file_contents(filename):
    try:
        with open(filename, "r") as file:
            data = file.read()
            print(f"--- Contents of {filename} ---")
            print(data)
    except FileNotFoundError:
        print("Error: The file was not found.")

read_file_contents("students.txt")
```

```python
#Q10 Handle division by zero using exception handling
def safe_divide(numerator, denominator):
    try:
        result = numerator / denominator
        print(f"{numerator} divided by {denominator} is: {result}")
    except ZeroDivisionError:
        print("Cannot divide by Zero!")

print("Test 1")
safe_divide(10, 2)

print("\nTest 2:")
safe_divide(5, 0)
```

```python
#Q11 Create a Student class with name and marks
class Student:
    # Constructor to initialize attributes
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Method to display student information
    def display_details(self):
        print(f"Student Name: {self.name}")
        print(f"Marks Scored: {self.marks}\n")

# To create objects
student1 = Student("Vaaruni", 91)
student2 = Student("Medhansh", 92)

# To display them
print("Student Records:")
student1.display_details()
student2.display_details()
```
