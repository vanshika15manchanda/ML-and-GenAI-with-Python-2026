Question 2 : Create a function to calculate sum of first N natural numbers

```python
def natural_num_sum(n):
  sum = 0
  for i in range(1,n+1):
    sum = sum + i
  return sum
natural_num_sum(23)
```

Question 3 : Create a function to reverse a number

```python
def rev(n):
  rev = 0
  while n != 0 :
    ld = n % 10
    rev = rev * 10 + ld
    n //= 10
  return rev
rev(9876556)
```

Question 4 : Create a function to count digits of a number

```python
def count_digits(n) :
  cnt = 0
  while n != 0 :
    n //= 10
    cnt = cnt +1
  return cnt
count_digits(678372687)
```

Question 5 : Create a function to check palindrome number

```python
def ispalindrome(n):
  def rev_number(num):
    reversed_num = 0
    while num != 0 :
      last_digit = num % 10
      reversed_num = reversed_num * 10 + last_digit
      num //= 10
    return reversed_num

  if n == rev_number(n):
    return True
  else:
    return False

# Test cases
print(f"Is 121 a palindrome? {ispalindrome(121)}")
print(f"Is 123 a palindrome? {ispalindrome(123)}")
print(f"Is 9876556 a palindrome? {ispalindrome(9876556)}")
print(f"Is 1221 a palindrome? {ispalindrome(1221)}")
```

Question 7 : Create a function to generate Fibonacci series.

```python
def fibonacci_series(n):
  a, b = 0, 1
  series = []
  for _ in range(n):
    series.append(a)
    a, b = b, a + b
  return series

# Demonstrate the function
print(f"Fibonacci series up to 10 terms: {fibonacci_series(10)}")
print(f"Fibonacci series up to 5 terms: {fibonacci_series(5)}")
```

Question 9 : Create a text file and store student details

```python

def store_student_details():
  name = input("Enter student name: ")
  age = input("Enter student age: ")
  grade = input("Enter student grade: ")

  with open("student_details.txt", "w") as file:
    file.write(f"Name: {name}\n")
    file.write(f"Age: {age}\n")
    file.write(f"Grade: {grade}\n")

  print("Student details saved to student_details.txt")

store_student_details()
```

Question 11 : Handle division by zero using exception handling

Question 12 : Create a student class with name and marks

```python
class Student:
  def __init__(self, name, marks):
    self.name = name
    self.marks = marks

  def display_student(self):
    print(f"Student Name: {self.name}, Marks: {self.marks}")


student1 = Student("Aiza", 85)
student1.display_student()

student2 = Student("shreya", 92)
student2.display_student()
```
