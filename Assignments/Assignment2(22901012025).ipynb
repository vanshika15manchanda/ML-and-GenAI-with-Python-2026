```python
#Finding sum of first 10 natural numbers
sum = 0 
for i in range(1, 11):
    sum += i
print("The sum of first 10 natural numbers is:", sum)
```

```python
#Finding factorial of number
n=int(input("Enter a number to find its factorial: "))
fact= 1
for i in range(1, n + 1):
    fact*= i
print("The factorial of", n, "is:", fact)
```

```python
#Fibonacci Series
n=int(input("Enter the number of terms in Fibonacci series: "))
a, b = 0, 1
print("Fibonacci series:")
for i in range(n):
    print(a, end=' ')
    a, b = b, a + b
```

```python
#Largest of three numbers
n1 = float(input("Enter first number: "))
n2 = float(input("Enter second number: "))
n3 = float(input("Enter third number: "))
if (n1 >= n2) and (n1 >= n3):
    largest = n1
elif (n2 >= n1) and (n2 >= n3):
    largest = n2
else:
    largest = n3

print("The largest number is:", largest)
```

```python
#Student Result System

#inputting details
name= input("Enter student's name: ")
roll_no= input("Enter student's roll number: ")
math= float(input("Enter marks obtained in Mathematics: "))
science= float(input("Enter marks obtained in Science: "))
english= float(input("Enter marks obtained in English: "))

#calculating percentage and display grade using if-else
total_marks = math + science + english
percentage = (total_marks / 500) * 100
grade=""
if percentage >= 90:
    grade = 'A'
elif percentage >= 80:
    grade = 'B'
elif percentage >= 70:
    grade = 'C'
elif percentage >= 60:
    grade = 'D'
else:
    grade = 'F'

print("\nStudent Result:", grade)
```
