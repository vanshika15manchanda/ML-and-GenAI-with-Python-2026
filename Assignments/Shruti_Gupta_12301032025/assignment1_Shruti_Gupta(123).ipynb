```python
#2. find simple interest
p=float(input("enter principal amount: "))
rate=float(input("enter rate without percentage: "))
r=rate*0.01
t=float(input("enter time period in years"))
i=(p*r*t)/100
print("simple interest: ",i)
```

```python
#7. Create a Student Report Program that take student details using input(), Store marks in variables,
#Calculate total and percentage , Use comments,Use proper indentation

name=input("Enter student name: ")
roll_no=input("Enter roll number: ")

#Input marks
math=float(input("Enter Math marks: "))
sci=float(input("Enter Science marks: "))
eng=float(input("Enter English marks: "))
# Total and percentage
total=math+sci+eng
percentage=(total/300)*100

# Display result
print("\n--- Student Report ---")
print("Name:", name)
print("Roll No:", roll_no)
print("Total Marks:", total,"out of 300.")
print("Percentage:", percentage, "%")
```
