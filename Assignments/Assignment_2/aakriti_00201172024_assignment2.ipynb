```python
# Program to calculate the sum of the first 10 natural numbers (1 to 10)
total_sum = 0

# Loop from 1 to 10
for i in range(1, 11):
    total_sum += i

print("The sum of the first 10 natural numbers is:", total_sum)
```

```python
# Program to find the factorial of a number using a loop
num = int(input("Enter a number to find its factorial: "))
factorial = 1

if num < 0:
    print("Factorial does not exist for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, num + 1):
        factorial *= i
    print(f"The factorial of {num} is {factorial}")
```

```python
# Program to print the Fibonacci series up to n terms
terms = int(input("How many terms of the Fibonacci series do you want? "))

# First two terms
n1, n2 = 0, 1
count = 0

if terms <= 0:
    print("Please enter a positive integer.")
elif terms == 1:
    print(f"Fibonacci series up to {terms} term: [0]")
else:
    print("Fibonacci series:")
    while count < terms:
        print(n1, end=" ")
        nth = n1 + n2
        # Update values
        n1 = n2
        n2 = nth
        count += 1
    print()  # For a clean newline at the end
```

```python
# Program to find the largest among three numbers using if-elif-else
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

if (num1 >= num2) and (num1 >= num3):
    largest = num1
elif (num2 >= num1) and (num2 >= num3):
    largest = num2
else:
    largest = num3

print("The largest number among the three is:", largest)
```

```python
# ==========================================
# Student Result System
# ==========================================

# 1. Input student details
student_name = input("Enter student name: ")
roll_no = input("Enter roll number: ")

# 2. Input marks for multiple subjects using a loop
num_subjects = int(input("Enter the number of subjects: "))
total_obtained_marks = 0
max_marks_per_subject = 100

for i in range(1, num_subjects + 1):
    while True:
        marks = float(input(f"Enter marks obtained in Subject {i} (out of 100): "))
        if 0 <= marks <= max_marks_per_subject:
            total_obtained_marks += marks
            break
        print("Invalid input! Marks should be between 0 and 100.")

# 3. Calculate percentage
total_max_marks = num_subjects * max_marks_per_subject
percentage = (total_obtained_marks / total_max_marks) * 100

# 4. Display grade using if-elif-else structural logic
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "Fail"

# Output execution Summary
print("\n==============================")
print("        FINAL RESULT          ")
print("==============================")
print("Name        :", student_name)
print("Roll No     :", roll_no)
print("------------------------------")
print(f"Total Marks : {total_obtained_marks:.1f} / {total_max_marks}")
print(f"Percentage  : {percentage:.2f}%")
print("Grade       :", grade)
print("==============================")
```
