**Find Sum of First 10 Natural Numbers**

```python
total = 0  # Initialize sum to 0

for i in range(1, 11):  # Loop from 1 to 10
    total = total + i

print("Sum of first 10 natural numbers =", total)
```

**Find Factorial of a Number**

```python
# Factorial of n = 1 × 2 × 3 × ... × n
# Example: factorial(5) = 1 × 2 × 3 × 4 × 5 = 120

num = int(input("Enter a number to find its factorial: "))

factorial = 1  # Initialize factorial to 1

if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("Factorial of 0 = 1")
else:
    for i in range(1, num + 1):  # Multiply from 1 to num
        factorial = factorial * i
    print(f"Factorial of {num} = {factorial}")
```

**Print Fibonacci Series**

```python
# Fibonacci Series: 0, 1, 1, 2, 3, 5, 8, 13, 21, ...

n = int(input("How many terms of Fibonacci series do you want? "))

# First two terms
a = 0
b = 1

print("Fibonacci Series:", end=" ")

for i in range(n):
    print(a, end=" ")   # Print current term
    next_term = a + b   # Calculate next term
    a = b               # Shift a to b
    b = next_term       # Shift b to next_term

print()  # New line at the end
```

**Find Largest Among 3 Numbers**

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

if num1 >= num2 and num1 >= num3:
    print(f"Largest number is: {num1}")
elif num2 >= num1 and num2 >= num3:
    print(f"Largest number is: {num2}")
else:
    print(f"Largest number is: {num3}")
```

**Student Result System**

```python
# ---------- Student Details ----------
name = input("Enter student name: ")
roll = input("Enter roll number: ")
cls  = input("Enter class: ")

# ---------- Marks Input ----------
maths    = float(input("Enter Maths marks (out of 100): "))
science  = float(input("Enter Science marks (out of 100): "))
english  = float(input("Enter English marks (out of 100): "))
hindi    = float(input("Enter Hindi marks (out of 100): "))
computer = float(input("Enter Computer marks (out of 100): "))

# ---------- Calculate Total and Percentage ----------
total      = maths + science + english + hindi + computer
max_marks  = 500
percentage = (total / max_marks) * 100

# ---------- Grade using if-elif-else ----------
if percentage >= 90:
    grade  = "A+"
    remark = "Outstanding!"
elif percentage >= 80:
    grade  = "A"
    remark = "Excellent!"
elif percentage >= 70:
    grade  = "B"
    remark = "Very Good!"
elif percentage >= 60:
    grade  = "C"
    remark = "Good"
elif percentage >= 40:
    grade  = "D"
    remark = "Pass"
else:
    grade  = "F"
    remark = "Fail"

# ---------- Display Result ----------
print("\n" + "="*45)
print("         STUDENT RESULT SYSTEM")
print("="*45)
print(f"  Name        : {name}")
print(f"  Roll No.    : {roll}")
print(f"  Class       : {cls}")
print("-"*45)
print(f"  Maths       : {maths}")
print(f"  Science     : {science}")
print(f"  English     : {english}")
print(f"  Hindi       : {hindi}")
print(f"  Computer    : {computer}")
print("-"*45)
print(f"  Total       : {total} / {max_marks}")
print(f"  Percentage  : {percentage:.2f}%")
print(f"  Grade       : {grade}")
print(f"  Remark      : {remark}")
print("="*45)
```
