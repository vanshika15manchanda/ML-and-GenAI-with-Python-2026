```python
# 1. Sum of First 10 Natural Numbers
sum_n = 0
for i in range(1, 11):
    sum_n += i
print("Sum of first 10 natural numbers =", sum_n)

# 2. Factorial of a Number
num = int(input("\nEnter a number to find factorial: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial =", fact)

# 3. Fibonacci Series
n = int(input("\nEnter number of terms for Fibonacci Series: "))
a, b = 0, 1
print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
print()

# 4. Largest Among 3 Numbers
n1 = float(input("\nEnter first number: "))
n2 = float(input("Enter second number: "))
n3 = float(input("Enter third number: "))

if n1 >= n2 and n1 >= n3:
    largest = n1
elif n2 >= n1 and n2 >= n3:
    largest = n2
else:
    largest = n3

print("Largest number =", largest)

# 5. Student Result System
print("\n----- Student Result System -----")

name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")

# Input marks using loop
total = 0
subjects = 5

for i in range(1, subjects + 1):
    marks = float(input(f"Enter marks of Subject {i}: "))
    total += marks

percentage = total / subjects

# Grade using if-elif-else
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

# Display Result
print("\n----- Result -----")
print("Name:", name)
print("Roll No:", roll_no)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
print("Grade:", grade)
```
