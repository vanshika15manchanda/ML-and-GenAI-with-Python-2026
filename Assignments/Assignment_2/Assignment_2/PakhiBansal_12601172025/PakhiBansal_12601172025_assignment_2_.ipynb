```python
#question no 5
name = str(input("enter your name :"))
class_section = int(input("enter you class: "))
maths = int(input("enter your maths marks: "))
phy = int(input("enter your physics marks: "))
chem = int(input("enter your chemistry marks: "))
bio = int(input("enter your biology marks: "))

marks = [maths,phy,chem,bio]
total = sum(marks)
percent = (total/400)*100
print(name)
print(class_section)
if percent >= 90:
    print(f"Your percentage is: {percent}% Grade A")
elif percent >= 80:
    print(f"Your percentage is: {percent}% Grade B")
elif percent >= 70:
    print(f"Your percentage is: {percent}% Grade C")
else:
    print(f"Your percentage is: {percent}% Grade D")

print("Thank you. All the best for future")
```
