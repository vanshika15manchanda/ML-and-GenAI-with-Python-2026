```python
#Q1 Find sum of first 10 natural no.s
sum=0
for i in range (11):
  sum+=i
print("Sum of first 10 natural no.s: ",sum)
```

```python
#Q2 Find factorial of a no.
num=int(input("Enter a number: "))
fact=1
for i in range(1,num+1):
  fact*=i
print("Factorial: ",fact)
```

```python
#Q3 Print Fibonnaci Series
n=int(input("Enter no. of terms(>1)"))
a=0
b=1
print(a)
print(b)
for i in range(0,n-2):
  c=b
  b=a+b
  a=c
  print(b)
```

```python
#Q5 Create a Student Report that takes student details,stores marks,calculates total and percentage

#Enter student details
name=input("Enter Name: ")
roll=int(input("Enter Roll no: "))

#Enter marks
eng_marks=int(input("Enter marks in English(/100): "))
math_marks=int(input("Enter marks in Maths(/100): "))
science_marks=int(input("Enter marks in Science(/100): "))
history_marks=int(input("Enter marks in History(/100): "))
french_marks=int(input("Enter marks in French(/100): "))

#Calculate total and percentage
total=eng_marks+math_marks+science_marks+history_marks+french_marks
per=total/5
grade=""
if per>=90:
  grade='A'
elif per>=80:
  grade='B'
elif per>=70:
  grade='C'
elif per>=60:
  grade='D'
else:
  grade='F'

#Display Report
print("---Report Card---")
print("Name: ",name,"\t Roll No: ",roll)
print("English:     ",eng_marks)
print("Maths  :     ",math_marks)
print("Science:     ",science_marks)
print("History:     ",history_marks)
print("French:      ",french_marks)
print("Total(/500): ",total)
print("Percantage : ",per,"%")
print("Grade :      ",grade)
```
