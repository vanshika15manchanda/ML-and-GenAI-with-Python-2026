```python
# 1 Find Sum of First 10 Natural Numbers
sum = 0
for i in range(1, 11):
    sum = sum + i
print("Sum =", sum)
```

```python
# 2 Find Factorial of a Number
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact = fact * i
print("Factorial =", fact)
```

```python
# 3 Print Fibonacci Series
n = int(input("Enter number of terms: "))
a = 0
b = 1
for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```
