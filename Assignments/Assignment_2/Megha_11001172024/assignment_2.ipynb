```python
#5. Create Student Result System
name = input("Enter Student Name: ")

total = 0

for i in range(1, 6):
    mark = int(input(f"Enter marks of Subject {i}: "))
    total += mark

percentage = total / 5

print("\nStudent Name:", name)
print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 90:
    print("Grade: A+")
elif percentage >= 80:
    print("Grade: A")
elif percentage >= 70:
    print("Grade: B")
elif percentage >= 60:
    print("Grade: C")
elif percentage >= 40:
    print("Grade: D")
else:
    print("Grade: F")
```
