```python
#QUESTION 2
#Find factorial of a no.

n = int(input("Enter the no.: "))

def fact(n):
    if n == 0 or n == 1:
        factorial = 1
        return factorial
    else:
        factorial = n * fact(n - 1)
        return factorial

print(f"{n}!= {fact(n)}")
```

```python
#QUESTION 4
#Find largest among 3 no.s

#Using built-in function
a = 2
b = 3
c = 5

largest = max(a, b, c)
print(f"Largest no.(using built in function is): {largest}")

#Without using built in function
a = 2
b = 3
c = 5

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print(f"Largest no.(without using built function is) : {largest}")
```

```python
#QUESTION 5
#Create Student Result System
# Input student details
# Input marks
# Calculate percentage
# Display grade
# Use:
# if-elif-else
# loops

print("----STUDENT DETAILS----")
name=str(input("Enter your Name: "))

subjects= 3
marks_list=[]
total=0

for i in range(1,subjects+i):
  marks=float(input(f"Enter marks of subject {i}: "))
  marks_list.append(marks)
  total=total+marks

percentage=total/subjects

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

print("\n----- RESULT -----")
print("Name:", name)
print("Marks:", marks_list)
print("Total:", total)
print("Percentage:", percentage)
print("Grade:", grade)
```
