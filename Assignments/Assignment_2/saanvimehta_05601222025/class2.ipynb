```python
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")
total = 0
for i in range(1, 4):
    marks = float(input(f"Enter marks in Subject {i}: "))
    total += marks
percentage = total / 3
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"
print("Name:", name)
print("Roll Number:", roll_no)
print("Total Marks:", total)
print("Percentage:", percentage)
print("Grade:", grade)
```
