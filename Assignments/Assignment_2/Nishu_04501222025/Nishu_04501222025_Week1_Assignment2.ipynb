```python
#1.Sum of first 10 natural numbers
sum_num = 0
for i in range(1,11):
  sum_num += i
print("Sum of first 10 natural numbers:",sum_num)
```

```python
#2.Factorial of a number
num = int(input("Enter a number:"))
fact = 1
for i in range(1,num+1):
  fact *= i
print("Factorial of",num,"is",fact)
```

```python
#3.Fibonacci Series
n = int(input("Enter the number of terms:"))
a,b = 0,1
print("Fibonacci Series:")
for i in range(n):
  print(a,end=" ")
  a,b = b,a+b
```

```python
#4.Largest among three numbers
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
num3 = int(input("Enter third number:"))
if num1 >= num2 and num1 >= num3:
  largest = num1
  print(num1,"is the largest number")
elif num2 >= num1 and num2 >= num3:
  largest = num2
  print(num2,"is the largest number")
else:
  largest = num3
  print(num3,"is the largest number")
print("The largest among 3 numbers is",largest)
```

```python
#5.Student Result System
name = input("Enter student name:")
roll_no = input("Enter student roll number:")
marks1 = int(input("Enter marks of English:"))
marks2 = int(input("Enter marks of Mathematics:"))
marks3 = int(input("Enter marks of Physics:"))
marks4 = int(input("Enter marks of Chemistry:"))
marks5 = int(input("Enter marks of Painting :"))
total_marks = marks1 + marks2 + marks3 + marks4 + marks5
percentage = (total_marks/500)*100
if percentage >= 90:
  grade = "A"
elif percentage >= 80:
  grade = "B"
elif percentage >= 70:
  grade = "C"
elif percentage >= 60:
  grade = "D"
elif percentage >= 40:
  grade = "E"
else:
  grade = "F"
print("Student Result:")
print("Student Name:",name)
print("Student Roll Number:",roll_no)
print("Subject 1 Marks:",marks1)
print("Subject 2 Marks:",marks2)
print("Subject 3 Marks:",marks3)
print("Subject 4 Marks:",marks4)
print("Subject 5 Marks:",marks5)
print("Total Marks:",total_marks)
print("Percentage:",percentage)
print("Grade:",grade)
```
