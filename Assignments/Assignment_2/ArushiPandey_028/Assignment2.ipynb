```python
#Question 1
sum=0
for i in range(1,11):
  sum+=i
print("Sum of first 10 natural numbers: ", sum)
```

```python
#Question 2
num=int(input("Enter a number: "))
fac=1
for i in range(1,num+1):
  fac*=i
print("Factorial of ", num, ": ", fac)
```

```python
#Question 3
num=int(input("Enter a number: "))
fib_1=0
fib_2=1
#Fibonacci Series
for i in range(1,num+1):
  print(fib_1, end=" ")
  next_term=fib_1+fib_2
  fib_1=fib_2
  fib_2=next_term
```

```python
#Question 4
num1=int(input("Enter a number: "))
num2=int(input("Enter a number: "))
num3=int(input("Enter a number: "))
if(num1>=num2 and num1>=num3):
    print(num1)
elif(num2>=num1 and num2>=num3):
    print(num2)
else:
    print(num3)
```

```python
#Question 5
#Taking student details as input
name=input("Enter student name: ")
rollNo = input("Enter Roll Number: ")
#Taking marks in 5 subjects as input and calculating total marks
totalMarks=0
for i in range(1,6):
  marks=int(input(f"Enter marks in subject {i}: "))
  totalMarks+=marks
#Calculating percentage
percentage=(totalMarks/500)*100
#Calculating grade
if(percentage>=80):
  grade="A"
elif(percentage>=60):
  grade="B"
elif(percentage>=40):
  grade="C"
elif(percentage>=20):
  grade="D"
else:
 grade="E"
#Printing the Student Report
print("Student Report:")
print("Student Name: ", name)
print("Roll Number:", rollNo)
print("Total Marks: ", totalMarks)
print("Percentage: ", percentage)
print("Grade: ", grade)
```
