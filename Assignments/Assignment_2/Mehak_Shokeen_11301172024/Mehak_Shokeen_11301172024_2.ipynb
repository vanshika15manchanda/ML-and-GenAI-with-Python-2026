##  1. Find Sum of First 10 Natural Numbers

```python
total = 0
# Loop from 1 to 10 (inclusive)
for i in range(1, 11):
    total = total + i  # Add each number to total

print("Sum of first 10 natural numbers =", total)
```


##  2. Find Factorial of a Number

```python
# Factorial of n = 1 × 2 × 3 × ... × n
num = int(input("Enter a number to find its factorial: "))
factorial = 1
for i in range(1, num + 1):
    factorial = factorial * i  # Multiply each number

print(f"Factorial of {num} = {factorial}")
```


##  3. Print Fibonacci Series

```python
# Fibonacci: 0, 1, 1, 2, 3, 5, 8, 13, ...
# Each number = sum of the two numbers before it
n = int(input("Enter number of terms for Fibonacci series: "))
a = 0
b = 1
print("Fibonacci Series:", end=" ")
for i in range(n):
    print(a, end=" ")
    next_term = a + b
    a = b
    b = next_term

print()
```


##  4. Find Largest Among 3 Numbers

```python
num1 = float(input("Enter first number : "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number : "))

# Compare all three numbers using if-elif-else
if num1 >= num2 and num1 >= num3:
    largest = num1
elif num2 >= num1 and num2 >= num3:
    largest = num2
else:
    largest = num3

print(f"Largest among {num1}, {num2}, {num3} = {largest}")
```


##  5. Student Result System
*(Uses: user input, loops, if-elif-else)*

```python
# Student Details (user input)
print("========== Student Result System ==========")
name        = input("Enter Student Name  : ")
roll_no     = input("Enter Roll Number   : ")
student_class = input("Enter Class         : ")

# Subject list
subjects = ["Mathematics", "Science", "English", "Hindi", "Social Science"]
marks_list = []

# Input marks using a loop
print("\nEnter marks for each subject (out of 100):")
for subject in subjects:
    mark = float(input(f"  {subject:<20}: "))
    marks_list.append(mark)

# Calculate total and percentage
total_marks = 0
for m in marks_list:
    total_marks = total_marks + m

max_marks  = len(subjects) * 100    # 5 subjects × 100 = 500
percentage = (total_marks / max_marks) * 100

# Determine grade using if-elif-else
if percentage >= 90:
    grade  = "A+"
    result = "PASS"
elif percentage >= 80:
    grade  = "A"
    result = "PASS"
elif percentage >= 70:
    grade  = "B"
    result = "PASS"
elif percentage >= 60:
    grade  = "C"
    result = "PASS"
elif percentage >= 40:
    grade  = "D"
    result = "PASS"
else:
    grade  = "F"
    result = "FAIL"

# Display Report Card
print("\n===========================================")
print("           STUDENT RESULT CARD            ")
print("===========================================")
print(f"  Name        : {name}")
print(f"  Roll No     : {roll_no}")
print(f"  Class       : {student_class}")
print("-------------------------------------------")

# Display each subject and its marks using loop
for i in range(len(subjects)):
    print(f"  {subjects[i]:<20}: {marks_list[i]}")

print("-------------------------------------------")
print(f"  Total Marks : {total_marks} / {max_marks}")
print(f"  Percentage  : {percentage:.2f}%")
print(f"  Grade       : {grade}")
print(f"  Result      : {result}")
print("===========================================")
```
