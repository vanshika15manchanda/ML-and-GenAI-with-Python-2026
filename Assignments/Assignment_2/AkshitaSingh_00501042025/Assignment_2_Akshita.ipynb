```python
#1.Find sum of first 10 natural numbers
sum = 0
for i in range (1,11):
  sum = sum + i
print("sum of first 10 natural numbers is :", sum)
```

```python
#2. Find factorial of a number
number=int(input("enter a number:"))
fact = 1
for i in range (1,number+1):
  fact = fact*i
print("Factorial of the number is :", fact)
```

```python
#3. Print fibonacci series
num = int(input("Enter number of terms : "))
a=0
b=1
print("Fibonacci series: ")

for i in range (num):
  print(a, end=" ")
  c= a+b
  a=b
  b=c
```

```python
#4. Find largest among 3 numbers
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
c=int(input("Enter third number: "))

if a>=b and a>=c:
  print(a, "is largest number")
elif b>=a and b>=c:
  print(b, "is largest number")
else:
  print(c, "is largest number")
```

```python
#5. Create student result system
name = input("Enter your name : ")
rollNo = input("Enter your roll number : ")
a = int(input("Enter marks of subject 1 : "))
b = int(input("Enter marks of subject 2 : "))
c = int(input("Enter marks of subject 3 : "))
d = int(input("Enter marks of subject 4 : "))
e = int(input("Enter marks of subject 5 : "))

total = a+b+c+d+e
percentage = (total/500)*100

if percentage>=90:
  Grade = "A"
elif percentage>=80:
  Grade = "B"
elif percentage>=70:
  Grade = "C"
elif percentage>=60:
  Grade = "D"
else:
  Grade = "FAIL"

print("Name : ", name)
print("Roll No : ", rollNo)
print("Total Marks : ", total)
print("Percentage : ", percentage)
print("Grade : ", Grade)
```
