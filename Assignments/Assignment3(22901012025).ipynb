```python
#print first 10 natural numbers
def natural(n):
    for i in range(1, n+1):
        print(i, end=' ')
natural(10)
```

```python
#calculate sum of first N natural numbers
def sum_natural(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum
n=10
```

```python
#reverse a number
def reverse_number(num):
    rev = 0
    while num > 0:
        rev = rev * 10 + num % 10
        num //= 10
    return rev
num = 314785
```

```python
#count digits in a number
def count_digits(num):
    count = 0
    while num > 0:
        count += 1
        num //= 10
    return count
num = 12345
```

```python
#number is palindrome
def is_palindrome(n):
    x = n
    r = 0
    while x > 0:
        r = r * 10 + x % 10
        x //= 10
    return r == n

n = 12321
if is_palindrome(n):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Fibonacci series up to N terms
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=' ')
        a, b = b, a + b
n = 10
fibonacci(n)
```

```python
#Calculator
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b != 0:
        return a / b
    return "Error! Division by zero."

print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

c = input("Enter choice (1/2/3/4): ")

n1 = float(input("Enter first number: "))
n2 = float(input("Enter second number: "))

if c == '1':
    r = add(n1, n2)
elif c == '2':
    r = sub(n1, n2)
elif c == '3':
    r = mul(n1, n2)
elif c == '4':
    r = div(n1, n2)
else:
    r = "Invalid input"

print("Result:", r)
```
