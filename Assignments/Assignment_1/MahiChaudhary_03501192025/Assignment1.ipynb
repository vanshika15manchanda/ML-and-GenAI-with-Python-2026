1. Area of rectangle

```python
length = float(input("Enter length: "))
breadth = float(input("Enter breadth: "))

area = length * breadth

print("Area of rectangle:", area)
```

2. Simple Interest

```python
p = float(input("Enter principal: "))
r = float(input("Enter rate: "))
t = float(input("Enter time: "))

si = (p*r*t)/100

print("Simple Interest:", si)
```

3. Convert Celsius to Fahrenheit

```python
# Program to convert Celsius to Fahrenheit

celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print("Temperature in Fahrenheit =", fahrenheit)
```

4. Average of 3 numbers

```python
# Program to calculate average of three numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

average = (a + b + c) / 3

print("Average =", average)
```

5. Square and Cube of a number

```python
# Program to find square and cube

num = int(input("Enter number: "))

square = num ** 2
cube = num ** 3

print("Square =", square)
print("Cube =", cube)
```

6. Swap two numbers without third variable

```python
# Program to swap two numbers without third variable

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a = a + b
b = a - b
a = a - b

print("After swapping")
print("a =", a)
print("b =", b)
```

7. Student Report Program

```python
# Program to create student report

name = input("Enter student name: ")
roll = input("Enter roll number: ")

sub1 = int(input("Enter marks 1: "))
sub2 = int(input("Enter marks 2: "))
sub3 = int(input("Enter marks 3: "))

total = sub1 + sub2 + sub3
percentage = total / 3

print("-----Student Report-----")
print("Name:", name)
print("Roll:", roll)
print("Total:", total)
print("Percentage:", percentage)
```
