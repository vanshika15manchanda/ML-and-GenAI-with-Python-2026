```python
length=int(input("Enter the length of the rectangle:"))
breadth=int(input("Enter the breadth of the rectangle:"))

area= length * breadth

print("The area of this rectangle is:", area)
```

```python
#Simple interest formula= (P*R*T)/100
p= int(input("Enter your principal rate:"))
r= int(input("Enter your interest rate:"))
t= int(input("Enter the time taken in years:"))
SI= (p*r*t)/100

print("Therefore, your Simple Interest is:",SI)
```

```python
#Converting Celsius To Fahrenheit = 9/5*(C)+32
c= float(input("Enter temperature in Celsius:"))
f= (1.8*c)+32
print("The given temperature in Fahrenheit:",f)
```

```python
#Average of 3 numbers
a=int(input("Enter the first number:"))
b=int(input("Enter the second number:"))
c=int(input("Enter the third number:"))
average=(a+b+c)/3

print("Average of the numbers:",average)
```

```python
#Finding square
a=int(input("Enter the number you want to square:"))
sq=a*a
print("The square of this number is",sq)

#Finding cube
b=int(input("Enter the number you'd like to cube:"))
cu=b*b*b
print("The cube of this number is",cu)
```
