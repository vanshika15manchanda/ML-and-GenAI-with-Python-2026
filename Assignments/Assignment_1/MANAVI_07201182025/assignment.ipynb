```python
#area of a rectangle
length = 10
breadth = 5
area = length * breadth
print("the area of the rectangle is:", area)

#find simple interest
principal = 500
rate = 2
time = 4
simple_interest = (principal * rate * time) / 100
print("the simple interest is:", simple_interest)

#converting temperature from Celsius to Fahrenheit
celsius = 50
fahrenheit = (celsius * 9/5) + 32
print("the temperature in Fahrenheit is:", fahrenheit)

#average of three numbers
a , b , c = 5,10,15
average = (a + b + c) / 3
print("the average of the three numbers is:", average)

#find sqauare and cube of a number
num = 6
square = num ** 2
cube = num ** 3
print("the square of the number is:", square)
print("the cube of the number is:", cube)

#Swapping two numbers without a third variable
x = 5
y = 10 
x = x + y
y = x - y       
x = x - y
print("after swapping, x is:", x)
print("after swapping, y is:", y)

#Student report program

#step 1: get name and roll number from user
name = input("enter the name of the student: ")
roll_number = input("enter the roll number of the student: ")

#step 2: get marks for subjects from user
maths = float(input("enter the marks for maths: "))
science = float(input("enter the marks for science: "))
english = float(input("enter the marks for english: "))

#step 3: calculate total marks and percentage
total_marks = maths + science + english
percentage = (total_marks / 300) * 100

print("student name:", name)
print("roll number:", roll_number)  
print("total marks:", total_marks)
print("percentage:", percentage)
```
