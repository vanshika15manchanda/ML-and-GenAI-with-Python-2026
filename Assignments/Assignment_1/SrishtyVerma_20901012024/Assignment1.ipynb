Q2. Find simple interest.

```python
P = (int) (input("Enter the Principal amount (in Rs) : "))
R = (int) (input("Enter the Rate of Interest (in %) : "))
T = (int) (input("Enter the Time (in yrs) : "))
simpleInterest = (P*R*T)/100
print("The Simple Interest as per the given data is: ", simpleInterest)
```

Q7. Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation

```python
#student details
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")

#marks as input
maths = float(input("Enter Maths marks (out of 100): "))
science = float(input("Enter Science marks (out of 100): "))
english = float(input("Enter English marks (out of 100): "))

#total marks
total = maths + science + english

#percentage
percentage = (total / 300)* 100

# Displaying the report
print("STUDENT REPORT : ")
print("Name:", name)
print("Roll Number:", roll_no)
print("Maths Marks:", maths)
print("Science Marks:", science)
print("English Marks:", english)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
