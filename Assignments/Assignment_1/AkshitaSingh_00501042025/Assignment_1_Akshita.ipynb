```python
#1. Find area of rectangle
length = float(input("Enter length of rectangle : "))
breath = float(input("Enter breath of rectangle : "))
area = length * breath
print("Area of rectangle =", area)
```

```python
#2. Find simple interest
p = float(input("Enter principal amount : "))
r = float(input("Enter rate of interest : "))
t = float(input("Enter time : "))
si = (p*r*t)/100
print("Simple interest =", si)
```

```python
#3. Convert temperature from celcius to Fahrenheit
celcius = float(input("Enter temperature in celcius : "))
fahrenheit = (celcius * 9/5) + 32
print("Temperature in Fahrenheit =", fahrenheit)
```

```python
#4. Calculate average of 3 numbers
a = float(input("Enter first number : "))
b = float(input("Enter second number : "))
c = float(input("Enter third number : "))
average = (a+b+c)/3
print("Average of three numbers =", average)
```

```python
#5.Find square and cube of a number
number = float(input("Enter a number : "))
square = number ** 2
cube = number ** 3
print("Square of a number =", square)
print("Cube of a number =", cube)
```

```python
#6. Swap 2 numbers without third variable
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))
a = a+b
b = a-b
a = a-b
print("After swapping, first number =", a)
print("After swapping, second number =", b)
```
