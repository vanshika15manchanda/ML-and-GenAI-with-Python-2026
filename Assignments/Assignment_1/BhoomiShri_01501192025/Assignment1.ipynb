```python
# 1.Find area of rectangle.
l=float(input("Enter the length of rectangle: "))
b=float(input("Enter the breadth of rectange: "))
area=l*b
print("Area of rectangle is :",area)
```

```python
# 2.Find simple interest.
p=float(input("Enter principal amount: "))
t=float(input("Enter time period(in years): "))
r=float(input("Enter rate of interest(in 2%): "))
I=(p*r*t)/100
print("Simple interest is :",I)
```

```python
# 3.Convert temperature from Celsius to Fahrenheit.
C=float(input("Enter temperature in celsius: "))
F=32+(9*C)/5
print("Temperature in fahrenheit: ",F)
```

```python
# 4.Calculate average of 3 numbers.
a=int(input("Enter 1st number : "))
b=int(input("Enter 2nd number : "))
c=int(input("Enter 3rd number : "))
avg= (a+b+c)/3
print("Average of numbers is :",avg)
```

```python
# 5.Find square and cube of a number.
n=float(input("Enter a number: "))
print("Square of number is :",n*n)
print("Cube of number is :",n*n*n)
```
