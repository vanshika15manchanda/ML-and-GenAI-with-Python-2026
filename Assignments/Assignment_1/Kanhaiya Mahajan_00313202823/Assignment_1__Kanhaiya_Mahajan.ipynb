```python
# area of Rectangle
# Input from user
length = float(input("Enter length: "))
breadth = float(input("Enter breadth: "))

# Calculate area
area = length * breadth

# Output result
print("Area of rectangle is:", area)
```

```python
# simple interest
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest: "))
time = float(input("Enter time in years: "))

# calculate the simple interest
simple_interest = (principal * rate * time) / 100

# print the result
print("Simple interest is:", simple_interest)
```

```python
# celcius to fahrenheit
celcius_temperature = float(input("Enter the celcius temperature"))
# convert tempetature from celcius to fahrenheit
fahrenheit_temperature = (celcius_temperature * 9/5) + 32
print("The fahrenheit temperature is:", fahrenheit_temperature)
```

```python
# Avg of three numbers
# input 3 numbers from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
# Calculate the average
avg = (num1 + num2 + num3) / 3
print("The average is:", avg)

```

```python
# find square and cube of a number
number = float(input("Enter a number: "))
square = number ** 2
cube = number ** 3

print("The square of", number, "is", square)
print("The cube of", number, "is", cube)
```

```python
# swap two numbers withour third variable
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))
# swapping a and b
a = a + b
b = a - b
a = a - b

print("After swapping:")
print("a =", a)
print("b =", b)

```

```python
# take student details and marks as input and print all the details with total marks and perecentage
student_branch = input("Enter the branch of the student: ")
student_name = input("Enter the name of the student: ")
student_roll_number = input("Enter the roll number of the student: ")
student_marks_subject1 = float(input("Enter the marks of subject 1 : "))
student_marks_subject2 = float(input("Enter the marks of subject 2 : "))
student_marks_subject3 = float(input("Enter the marks of subject 3 : "))

total_marks = student_marks_subject1 + student_marks_subject2 + student_marks_subject3
percentage = (total_marks / 300) * 100

print("Branch:", student_branch)
print("Name:", student_name)
print("Roll Number:", student_roll_number)
print("Marks in Subject 1:", student_marks_subject1)
print("Marks in Subject 2:", student_marks_subject2)
print("Marks in Subject 3:", student_marks_subject3)
print("Total Marks:", total_marks)
print("Percentage:", percentage)


```
