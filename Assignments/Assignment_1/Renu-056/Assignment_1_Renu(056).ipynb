Q-2 Find simple interest.

```python
amount=int(input("enter the amount:"))
time_need=int(input("enter the given time period in years:"))
rate=int(input("enter the rate:"))

simple_interst=(amount*time_need*rate)/100

print(f"your interest is {simple_interst}")
```

Q-3 Convert temperature from celcius to fahrenheit.

```python
temp_in_celcius=int(input("Temperature in celcius:"))
convertor=(temp_in_celcius*1.8)+32

print(f"The given temperature in fahrenheit is {convertor} degree fahrenheit.")
```
