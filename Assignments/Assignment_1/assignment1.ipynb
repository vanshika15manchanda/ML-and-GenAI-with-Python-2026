```python
#area of rectangle
a=float(input("Length of rectangle: "))
b=float(input("Breadth of rectangle: "))
print("area of rectangle is: ",a*b)
```

```python
#simple interest
p=float(input("Enter principal amount: "))
r=float(input("Enter Rate of Interest: "))
t=float(input("Enter time: "))
SI=(p*r*t)/100
print(SI)
```

```python
#celsius to fahrenheit
c=float(input("Enter temperature in celsius: "))
f=(c*1.8)+32
print("Temperature in fahrenheit is:",f)
```

```python
#average of three numbers
a=float(input("Enter first number: "))
b=float(input("Enter second number: "))
c=float(input("Enter third number: "))
avg=(a+b+c)/3
print("The average of three numbers is:",avg)
```

```python
#square and cube of a number
a=float(input("Enter a number: "))
print("The square of the number is:",a**2)
print("The cube of the number is:",a**3)

```

```python
#swapping two numbers
a=float(input("Enter first number: "))
b=float(input("Enter second number: "))
a,b=b,a
print("After swapping:")
print("First number:",a)
print("Second number:",b)
```

```python
#student report program
name=input("Enter name of student: ")
roll=int(input("Enter roll number: "))
#enter marks
english=float(input("Enter marks in english: "))
hindi=float(input("Enter marks in hindi: "))
maths=float(input("Enter marks in maths: "))
science=float(input("Enter marks in science: "))
total= english + hindi + maths + science
marks_percentage= total/4
print("\n STUDENT REPORT")
print("Name of student:",name)
print("Roll number:",roll)
print("Total marks attained:",total)
print("Percentage attained:",marks_percentage,"%")
```
