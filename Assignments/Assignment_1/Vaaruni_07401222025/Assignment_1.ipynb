```python
#Q1 Find Area of a rectangle
l=int(input("Enter area of the rectangle: "))
b=int(input("Enter breadth of the rectangle: "))
area=l*b
print("Area of the rectangle: ",area)
```

```python
#Q2 Find Simple interest
p=int(input("Enter Principle amount: "))
r=int(input("Enter rate: "))
t=int(input("Enter time: "))
s=(p*r*t)/100
print("Simple Interest is: ",s)
```

```python
#Q3 Convert temperature from Celsius to Fahrenheit
c=int(input("Enter Temp in Celsius: "))
f=(9/5)*c+32
print("Temp in Fahrenheit: ",f)
```

```python
#Q4 Calculate average of three numbers
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
c=int(input("Enter third number: "))
avg=(a+b+c)/3
print("Average of three numbers is :",avg)
```

```python
#Q5 Find square and cube of a number
x=int(input("Enter a number: "))
print("Square of the number",x**2)
print("Cube of the number",x**3)
```

```python
#Q6 Swap 2 numbers without a third variable
a=int(input("Enter a number: "))
b=int(input("Enter another number: "))
a=a+b
b=a-b
a=a-b
print("Variable a: ",a,"\nVariable b: ",b)
```

```python
#Q7 Create a Student Report that takes student details,stores marks,calculates total and percentage

#Enter student details
name=input("Enter Name: ")
roll=int(input("Enter Roll no: "))

#Enter marks
eng_marks=int(input("Enter marks in English(/100): "))
math_marks=int(input("Enter marks in Maths(/100): "))
science_marks=int(input("Enter marks in Science(/100): "))
history_marks=int(input("Enter marks in History(/100): "))
french_marks=int(input("Enter marks in French(/100): "))

#Calculate total and percentage
total=eng_marks+math_marks+science_marks+history_marks+french_marks
per=total/5

#Display Report
print("---Report Card---")
print("Name: ",name,"\t Roll No: ",roll)
print("English:     ",eng_marks)
print("Maths  :     ",math_marks)
print("Science:     ",science_marks)
print("History:     ",history_marks)
print("French:      ",french_marks)
print("Total(/500): ",total)
print("Percantage : ",per,"%")
```
