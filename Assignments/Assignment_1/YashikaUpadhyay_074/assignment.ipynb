Exercise 1

```python
# input the length of the rectangle from the user
length = float(input("Enter length: "))

# input the breadth of the rectangle from the user
breadth = float(input("Enter breadth: "))

# calculating area of the rectangle
area = length * breadth

# display the calculated result
print("Area of rectangle =", area)
```

Exercise 2

```python
# input the principal amount from the user
principal = float(input("Enter principal amount: "))

# input rate of interest from the user
rate = float(input("Enter rate of interest: "))

# input time in years from the user
time = float(input("Enter time (in years): "))

# calculating simple interest
simple_interest = (principal * rate * time) / 100

# display the calculated result
print("Simple Interest =", simple_interest)
```

Exercise 3

```python
# input temperature in celsius from the user
celsius = float(input("Enter temperature in Celsius: "))

# calculate the temperature in fahrenheit
fahrenheit = (celsius * 9/5) + 32

# display the calculated result
print("Temperature in Fahrenheit =", fahrenheit)
```

```python
# input the numbers from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# calculate the average of the three numbers
average = (num1 + num2 + num3) / 3

# display the calculated result
print("Average =", average)
```

```python
# input the number from the user
num = int(input("Enter a number: "))

# calculate the square and cube of the number
square = num ** 2
cube = num ** 3

# display the calculated results
print("Square =", square)
print("Cube =", cube)
```

```python
# input the numbers from the user
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# steps to swapping the variables without a third variable
a = a + b
b = a - b
a = a - b

# display the calculate results
print("After swapping:")
print("a =", a)
print("b =", b)
```

Exercise 7

```python
# Input student details and marks
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")
marks1 = float(input("Enter marks in English: "))
marks2 = float(input("Enter marks in Mathematics: "))
marks3 = float(input("Enter marks in Science: "))

# Calculating total marks and percentage
total = marks1 + marks2 + marks3
percentage = (total / 300) * 100

# Displaying report
print("\n----- STUDENT REPORT -----")
print("Name       :", name)
print("Roll No.   :", roll_no)
print("English    :", marks1)
print("Mathematics:", marks2)
print("Science    :", marks3)
print("Total Marks:", total)
print("Percentage :", percentage, "%")
```
