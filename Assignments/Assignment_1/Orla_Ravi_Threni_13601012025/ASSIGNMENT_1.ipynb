```python
# 1. Area of Rectangle
l = float(input("Length: "))
b = float(input("Breadth: "))
print("Area =", l * b)
```

```python
# 2. Simple Interest
p = float(input("Principal: "))
r = float(input("Rate: "))
t = float(input("Time: "))
si = (p * r * t) / 100
print("Simple Interest =", si)
```

```python
# 3. Celsius to Fahrenheit
c = float(input("Celsius: "))
f = (c * 9/5) + 32
print("Fahrenheit =", f)
```

```python
# 4. Average of 3 Numbers
a = float(input())
b = float(input())
c = float(input())

avg = (a + b + c) / 3
print("Average =", avg)
```

```python
# 5. Square and Cube
n = int(input("Number: "))
print("Square =", n * n)
print("Cube =", n * n * n)
```

```python
# 6. Swap Without Third Variable
a = int(input("a: "))
b = int(input("b: "))

a, b = b, a

print("a =", a)
print("b =", b)
```

```python
# 7. Student Report Program
name = input("Name: ")

m1 = int(input("Mark 1: "))
m2 = int(input("Mark 2: "))
m3 = int(input("Mark 3: "))

total = m1 + m2 + m3
percentage = total / 3

print("Name:", name)
print("Total:", total)
print("Percentage:", percentage)
```
