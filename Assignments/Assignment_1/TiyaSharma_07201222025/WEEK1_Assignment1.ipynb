```python
#QUESTION 7
# Student Report Program

# Taking student details using input()
student_name = input("Enter student name: ")
roll_number = input("Enter roll number: ")

# Storing marks of three subjects in variables
math_marks = float(input("Enter Math marks: "))
science_marks = float(input("Enter Science marks: "))
english_marks = float(input("Enter English marks: "))

# Calculating total marks
total_marks = math_marks + science_marks + english_marks

# Calculating percentage (assuming each subject is out of 100)
maximum_possible_marks = 300
percentage = (total_marks / maximum_possible_marks) * 100

# Displaying the student report card
print("\n--- Student Report Card ---")
print(f"Name: {student_name}")
print(f"Roll Number: {roll_number}")
print(f"Total Marks Obtained: {total_marks} / {maximum_possible_marks}")
print(f"Percentage: {percentage:.2f}%")
```
