## 1. Find area of rectangle.

```python
length = int(input("Enter length :- "))
breadth = int(input("Enter breadth :- "))
area = length*breadth
print(f"Area of rectangle is {area}") 
```

## 2. Find simple interest.

```python
principle = float(input("Enter principle :- "))
rate = float(input("Enter rate :- "))
time = int(input("Enter time :- "))
SI = (principle*rate*time)/100
print(f"Simple Interest is {SI}")
```

## 3. Convert temperature from Celsius to Fahrenheit.

```python
cel = float(input("Enter temperature in celsius :- "))
far = (cel + 1.8) + 32
print(f"Temperature in Farenheit is {far}")
```

## 4. Calculate average of 3 numbers.

```python
num1 = float(input("Enter first number :- "))
num2 = float(input("Enter second number :- "))
num3 = float(input("Enter third number :- ")) 
avg = (num1 + num2 + num3)/3
print(f"Average of three numbers is {avg:.3f}")
```

## 5. Find square and cube of a number.

```python
n = int(input("Enter number :- "))
print(f"Square of number is :- {n**2}")
print(f"Cube of number is :- {n**3}")
```

## 6. Swap two numbers without third variable.

```python
a = int(input("Enter a :- ")) # a = 7
b = int(input("Enter b :- ")) # b = 5
print(f"Before Swapping\na = {a}\nb = {b}")
a = a + b
b = a - b
a = a - b
print(f"Before Swapping\na = {a}\nb = {b}")
```

## 7. Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation

```python
# To input student's name
name = input("Enter your name :- ")

# Entering marks of five different subjects
maths = float(input("\nEnter your marks in maths :- "))
science = float(input("Enter your marks in science :- "))
english = float(input("Enter your marks in english :- "))
social_science = float(input("Enter your marks in social science :- "))
hindi = float(input("Enter your marks in hindi :- "))

# Calculating total marks
total = maths + science + english + social_science + hindi

# Calculating percentage of student
percentage = (total/500)*100

# Printing total marks
print(f"\n{name}, your total marks are :- {total}")
print(f"{name}, your percentage is :- {percentage:.1f}")
```
