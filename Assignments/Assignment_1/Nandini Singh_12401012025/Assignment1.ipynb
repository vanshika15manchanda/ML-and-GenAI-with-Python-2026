```python
#ASSIGNMENT 1

#Q1
length=float(input("Enter the length: "))
breadth=float(input("Enter the breadth: "))
area=length*breadth
print("The area is: ",area)
```

```python
#Q2
principal=float(input("Enter Principal amount: "))
rate=float(input("Enter the rate of interest: "))
time=float(input("Enter the time period: "))
si=(principal*rate*time)/100

print("The simple interest is: ",si)
```

```python
#Q3
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32

print("Temperature in Fahrenheit =", fahrenheit)
```

```python
#Q4
num1=float(input("Enter first number: "))
num2=float(input("Enter second number: "))
num3=float(input("Enter third number: "))

average=(num1+num2+num3)/3
print("The average is: ",average)
```

```python
#Q5
a=float(input("Enter the number: "))
square=a*a

cube=a*a*a

print ("The square of the number is: ",square)
print ("The cube of the number is: ",cube)
```

```python
#Q6
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a = a + b
b = a - b
a = a - b

print("After Swapping:")
print("a =", a)
print("b =", b)
```

```python
# Q7
name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")

mark1 = float(input("Enter marks in Subject 1: "))
mark2 = float(input("Enter marks in Subject 2: "))
mark3 = float(input("Enter marks in Subject 3: "))

total = mark1 + mark2 + mark3

percentage = total / 3

print("\n STUDENT REPORT \n")
print("Name:", name)
print("Roll Number:", roll_no)
print("Subject 1 Marks:", mark1)
print("Subject 2 Marks:", mark2)
print("Subject 3 Marks:", mark3)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
