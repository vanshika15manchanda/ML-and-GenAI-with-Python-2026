```python
#1.Area of Rectangle
length = float(input("Enter the length of the rectangle:"))
width = float(input("Enter the width of the rectangle:"))
area = length * width
print("The area of the rectangle is:", area)
```

```python
#2.Simple Interest
principal = int(input("Enter the principal amount:"))
rate = int(input("Enter the rate of interest:"))
time = int(input("Enter the time period in years:"))
simple_interest = (principal * rate * time) / 100
print("The simple interest is:", simple_interest)
```

```python
#3.Celsius to Fahrenheit Conversion
celsius = int(input("Enter the temperature in Celsius:"))
fahrenheit = (celsius * 9/5) + 32
print("The temperature in Fahrenheit is:", fahrenheit)
```

```python
#4.Average of 3 Numbers
num1 = int(input("Enter the first number:"))
num2 = int(input("Enter the second number:"))
num3 = int(input("Enter the third number:"))
average = (num1 + num2 + num3) / 3
print("The average is:",average)
```

```python
#5.Square and Cube of a Number
number = int(input("Enter a number:"))
square = number ** 2
cube = number ** 3
print("The square of the number is:", square)
print("The cube of the number is:", cube)
```

```python
#6.Swap two Numbers Without Third Variable
a = int(input("Enter the first number:"))
b = int(input("Enter the second number:"))
print("Before swapping:")
print("a =", a)
print("b =", b)
a,b=b,a
print("After swapping:")
print("a =", a)
print("b =", b)
```

```python
#7.Student Report Program
#Taking student details
name = input("Enter the student's name:")
roll_number = input("Enter the student's roll number:")
#Taking marks
english = int(input("Enter marks for English:"))
maths = int(input("Enter marks for Maths:"))
science = int(input("Enter marks for Science:"))
#Calculating total and percentage
total = english + maths + science
percentage = (total / 300) * 100
#Printing the report
print("Student Report:")
print("Name:", name)
print("Roll Number:", roll_number)
print("Marks in English:", english)
print("Marks in Maths:", maths)
print("Marks in Science:", science)
print("Total Marks:", total)
print("Percentage:", percentage,"%")
```
