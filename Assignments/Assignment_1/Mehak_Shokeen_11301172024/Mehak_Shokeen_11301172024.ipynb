## 1. Find Area of Rectangle

```python
# Formula: Area = length × breadth

length = float(input("Enter the length of the rectangle: "))
breadth = float(input("Enter the breadth of the rectangle: "))
area = length * breadth

print(f"Area of Rectangle = {area}")
```

##  2. Find Simple Interest

```python
# Formula: SI = (Principal × Rate × Time) / 100

principal = float(input("Enter Principal amount: "))
rate = float(input("Enter Rate of Interest (in %): "))
time = float(input("Enter Time (in years): "))

simple_interest = (principal * rate * time) / 100

print(f"Simple Interest = {simple_interest}")
```


##  3. Convert Temperature from Celsius to Fahrenheit

```python
# Formula: F = (C × 9/5) + 32

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C = {fahrenheit}°F")
```


### 4. Calculate Average of 3 Numbers

```python
# Formula: Average = (a + b + c) / 3

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

average = (num1 + num2 + num3) / 3

print(f"Average of {num1}, {num2}, {num3} = {average}")
```


## 5. Find Square and Cube of a Number

```python
number = float(input("Enter a number: "))
square = number ** 2   # Square of the number
cube = number ** 3     # Cube of the number

print(f"Square of {number} = {square}")
print(f"Cube of {number}   = {cube}")
```


## 6. Swap Two Numbers Without Third Variable

```python
a = int(input("Enter first number (a): "))
b = int(input("Enter second number (b): "))
print(f"Before Swap: a = {a}, b = {b}")

# Swapping using arithmetic operations
a = a + b   # a now holds sum of both
b = a - b   # b gets original value of a
a = a - b   # a gets original value of b

print(f"After Swap:  a = {a}, b = {b}")
```


## 7. Student Report Program

```python
# Student Details
print("===== Student Report Card =====")
student_name = input("Enter Student Name: ")
roll_number  = input("Enter Roll Number: ")
student_class = input("Enter Class: ")

# Subject Marks Input
print("\nEnter marks for each subject (out of 100):")
marks_maths   = float(input("  Mathematics : "))
marks_science = float(input("  Science     : "))
marks_english = float(input("  English     : "))
marks_hindi   = float(input("  Hindi       : "))
marks_sst     = float(input("  Social Sc.  : "))

# Calculations
total_marks     = marks_maths + marks_science + marks_english + marks_hindi + marks_sst
max_marks       = 500                               # 5 subjects × 100
percentage      = (total_marks / max_marks) * 100   # Percentage formula

# Grade Assignment
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 40:
    grade = "D"
else:
    grade = "F (Fail)"

# Report Display
print("\n========================================")
print("           STUDENT REPORT CARD          ")
print("========================================")
print(f"  Name        : {student_name}")
print(f"  Roll No     : {roll_number}")
print(f"  Class       : {student_class}")
print("----------------------------------------")
print(f"  Mathematics : {marks_maths}")
print(f"  Science     : {marks_science}")
print(f"  English     : {marks_english}")
print(f"  Hindi       : {marks_hindi}")
print(f"  Social Sc.  : {marks_sst}")
print("----------------------------------------")
print(f"  Total Marks : {total_marks} / {max_marks}")
print(f"  Percentage  : {percentage:.2f}%")
print(f"  Grade       : {grade}")
print("========================================")
```
