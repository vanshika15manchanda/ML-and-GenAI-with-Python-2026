```python
# Taking student details using input()
student_name = input("Enter student name: ")
roll_number = input("Enter roll number: ")

# Store marks in variables
# Assuming 3 subjects for this example
math_marks = float(input("Enter marks for Mathematics: "))
science_marks = float(input("Enter marks for Science: "))
history_marks = float(input("Enter marks for History: "))

# Calculate total and percentage
total_marks = math_marks + science_marks + history_marks
# Assuming each subject is out of 100
percentage = (total_marks / 300) * 100

# Displaying the final report with proper indentation
print("\n---------- Student Report Card ----------")
print("Name:         ", student_name)
print("Roll Number:  ", roll_number)
print("-----------------------------------------")
print("Total Marks:  ", total_marks)
print("Percentage:   ", round(percentage, 2), "%")
print("-----------------------------------------")
```
