```python
#Question 7
#Taking student details as input
name=input("Enter student name: ")
rollNo = input("Enter Roll Number: ")
#Taking marks in 5 subjects as input
marks1=int(input("Enter marks in subject1: "))
marks2=int(input("Enter marks in subject2: "))
marks3=int(input("Enter marks in subject3: "))
marks4=int(input("Enter marks in subject4: "))
marks5=int(input("Enter marks in subject5: "))
#Calculating total marks
totalMarks=marks1+marks2+marks3+marks4+marks5
#Calculating percentage
percentage=(totalMarks/500)*100
#Printing the Student Report
print("Student Report:")
print("Student Name: ", name)
print("Roll Number:", rollNo)
print("Total Marks: ", totalMarks)
print("Percentage: ", percentage)
```
