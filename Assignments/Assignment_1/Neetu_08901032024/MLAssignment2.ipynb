```python
#Find factorial of a number
num = int(input("Enter a number: "))
fact= num
i = num -1
while i > 0:
    fact = fact*i
    i -= 1
print(fact)
```

```python
#Print fibonacci series
num = int(input("enter a number: "))
first= 0
second = 1
count = 0
while count< num:
    print(first)
    sum = first + second
    first = second
    second = sum
    count += 1
```

```python
#Create Student Result Svstem:Input student details,Input marks,Calculate percentage,Display grade
name = input("enter student name: ")
rno = int(input("enter roll no. : "))
total = 0
max_marks = 500
for i in range(1,6):
    marks = float(input(f" enter marks for subject {i}:"))
    total = total + marks
perc = (total/max_marks)*100
print("Student details: ")
print("Name: ",name)
print("Roll no.: ",rno)
print("Total marks obtained: ",total)
print("Percentage: ",perc)
print("Grade: ")
if perc>=90:
  print("Grade A")
elif perc>=80:
  print("Grade B")
elif perc>= 70:
  print("Grade C")
elif perc>=50:
  print("Grade D")
else:
  print("Fail")
```
