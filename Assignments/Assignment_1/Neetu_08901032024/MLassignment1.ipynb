```python
# find simple interest
P = int(input("Enter principal amount: "))
R = int(input("Enter annual interest rate: "))
T = int(input("Enter duration for which the money is borrowed: "))
I = (P*R*T)/100
print("Simple interest: ",I)
```

```python
#convert temperature from celsius to fahrenheit
C = int(input("Enter temperature in celsius: "))
F = (C*9/5)+32
print("Temperature in fahrenheit = ",F)
```

```python
#find square and cube of a number
num= int(input("Enter the number: "))
sq = num*num
cube = num*num*num
print("Square of number is ",sq)
print("Cube of number is ",cube)
```

```python
#swap two numbers without third variable
a= int(input("Enter number 1: "))
b=int(input("Enter number 2: "))
print("Before swap: ",a,",",b)
a = a+b
b= a-b
a= a-b
print("After swap: ",a,",",b)
```

```python
#create student report program
name= input("Enter name of student: ")
rno= int(input("Enter roll no. of student: "))
age= int(input("Enter age of student: "))
#collecting marks of students
marks1 = int(input("Enter marks of physics: "))
marks2 = int(input("Enter marks of OS: "))
marks3 = int(input("Enter marks of Devops: "))
marks4 = int(input("Enter marks of maths: "))
marks5 = int(input("Enter marks of DBMS: "))
max_marks= 500
#calculating total marks
total= marks1+ marks2+ marks3+ marks4+ marks5
#calculating percentage
perc = (total/max_marks)*100
#print students details
print("Student details")
print("Name of student: ",name)
print("Roll no.: ",rno)
print("Age: ",age)
print("Marks details")
print("Physics: ",marks1)
print("OS: ",marks2)
print("Devops: ",marks3)
print("Maths: ",marks4)
print("DBMS: ",marks5)
print("Total marks obtained: ",total)
print("Percentage: ",perc)
```
