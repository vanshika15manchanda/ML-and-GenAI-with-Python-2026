```python
#Find area of rectangle.

l=int(input("Enter length of the rectangle: "))
b=int(input("Enter breadth of the rectangle: "))
area=l*b
print("Area of a rectangle with length",l,"and breadth",b,"is",area)
```

```python
#Find simple interest.
#Formula of simple interest: (P*R*T)/100

p=float(input("Enter Principal Amount(initial amount): "))
r=float(input("Enter Interest Rate in percentage: "))
t=float(input("Enter Time in Years: "))
si=(p*r*t)/100
print("The simple interest with given values calculated is", si)
```

```python
#Convert temperature from Celsius to Fahrenheit.
#F = (C*1.8)+32

c=float(input("Enter temperature in Celsius: "))
f=(c*1.8)+32
print("Temperature in Fahrenheit: ", f)
```

```python
#Swap two numbers without third variable.
a=int(input("Enter a number: "))
b=int(input("Enter another number: "))
print("Current value of a: ", a)
print("Current value of b: ", b)
a,b=a-a+b,b-b+a
print("New value of a: ", a)
print("New value of b: ", b)
```

```python
#Create a Student Report Program that take student details using input, Store marks in variables, Calculate total and percentage, Use comments,
#Use proper indentation

#Input student details
name=input("Enter Student Name: ")
rno=input("Enter Student Roll Number: ")

#Input marks
sub1=float(input("Enter marks in Subject 1 (out of 100): "))
sub2=float(input("Enter marks in Subject 2 (out of 100): "))
sub3=float(input("Enter marks in Subject 3 (out of 100): "))
sub4=float(input("Enter marks in Subject 4 (out of 100): "))
sub5=float(input("Enter marks in Subject 5 (out of 100): "))

#total marks and percentage
total=sub1+sub2+sub3+sub4+sub5
percentage=(total/500)*100

print("\nStudent Report Program")
print("Student Name: ", name)
print("Roll Number: ", rno)
print("Total Marks: ", total)
print("Percentage: ", percentage,"%")
```
