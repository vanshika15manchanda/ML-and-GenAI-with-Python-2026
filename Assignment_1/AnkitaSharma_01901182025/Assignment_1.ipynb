```python
#1 Find Area of Rectangle
length= int(input("Enter the length of the rectangle:")) #takes the length of the rectangle as input
breadth= int(input("Enter the breadth of the rectangle:")) #takes the breadth of the rectangle as input
area= length*breadth #calculates the area
print("Area of the given rectangle is:",area) #prints the area as the output
```

```python
#2 Find simple interest
p= int(input("Enter the principal amount:")) #takes the principal amount as input
r= int(input("Enter the rate of interest:")) #takes the rate of interest as imput
t= int(input("Enter the time period in years:")) #takes the time period as input
print("The simple interest:",(p*r*t)/100)
```

```python
#3 Convert temperature from celsius to Fahrenheit
temp= int(input("Enter the temperature in celsius:")) #takes temperautre in celcius as input
f = temp*(9/5)+32 #converts celsius to fahrenheit
print("Temoerature in celcius:", f) # prints the output
```

```python
#4 Calculate average of three numbers
num1= int(input("Enter number 1:")) #takes first number as input
num2= int(input("Enter number 2:")) #takes second number as input
num3= int(input("Enter number 3:")) #takes third number as input
avg= (num1+num2+num3)/3 #calculates the average
print("Average of the numbers given:",avg) #prints the output
```

```python
#5 Find square and cube of a number
num= int(input("Enter the number: ")) #takes the number as input
print("Square of ", num,'is ',num**2,"\n","Cube of ", num, 'is ', num**3)
```

```python
#6 Swap two numbers without third variable
num1= int(input("Enter number 1:"))
num2= int(input("Enter number 2:"))
print("initially number 1:",num1,"and number 2:", num2)
num1, num2= num2, num1 #swapping
print("After swapping number 1:",num1,"and number 2:", num2)
```

```python
#7 Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage
detail= input("Enter student marks in dictionary: ") #taking input of all the marks, subject-wise
marks_dict = eval(detail) #converting the details into dictionary
marks= marks_dict.values() #getting a list of all marks
total= sum(marks) #calculating total
print("Total Marks: ",total, "\nPercentage:", total/(len(marks))) #final output
```
