```python
#simple interest
p_amt=float(input("Enter principal amt in Rs. "))
rate=float(input("Enter rate of interest in percentage (%) :"))
time=float(input("Enter time in years: "))
si=(p_amt*rate*time)/100
print("Simple interest is: ", "Rs.", si)
```
