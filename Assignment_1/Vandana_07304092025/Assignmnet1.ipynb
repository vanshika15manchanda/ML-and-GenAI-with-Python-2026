## 1. Find area of rectangle

```python
length = float(input("enter length of rectangle:"))
width = float(input("enter length of width:"))
area = length * width
print("Area of rectangle is",area)
```

## 2. Find Simple Interest


```python
pri = int(input("enter principle amount:"))
rate = int(input("enter rate of interest:"))
time = int(input("enter time:"))
si = (pri*rate*time)/100
print("Simple Interest is",si)
```

## 3. Covert temperature from celcius to fahrenheit

```python
celcius = float(input("enter temperature in celcius:"))
fah = (celcius * 9/5) + 32
print("temperature in fahrenheit is",fah)
```

## 4. calculate average of 3 numbers

```python
num1 = int(input("enter first number:"))
num2 = int(input("enter second number:"))
num3 = int(input("enter third number:"))
avg = (num1+num2+num3)/3
print("average of three numbers is",avg)
```

## 4. Find square and cube of a number

```python
num = int(input("enter a number:"))
square = num**2
cube = num**3
print("square of a number is",square)
print("cube of a number is",cube)
```

##5. Swap two numbers without third variable

## 6. create a student report program that take student details using input(), store marks in variables, calculate total percentage, use comments, use proper indentation

```python
name = input("enter student name:")
roll_no = int(input("enter roll number:"))
n = int(input("enter number of subjects:"))
marks = []
for i in range(n):
    mark = int(input("enter marks:"))
    marks.append(mark)
print("student report:")
print("student name:",name)
print("roll number:",roll_no)
print("marks:",marks)
print("percentage:",sum(marks)/n)
```
