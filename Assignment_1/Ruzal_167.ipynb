Find simple interest.

```python
p = int(input("Enter principle : "))
r = int(input("Enter rate of interest : "))
t = int(input("Enter time : "))
si = (p*r*t)/100
print(si)
```

Convert temperature from Celsius to Fahrenheit.

```python
celsius = float(input("Enter temperature in celsius : "))
fahrenheit = (celsius * 9/5) + 32
print(fahrenheit)
```

Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation

```python
name  = input("Enter name : ")
roll_no = input("Roll No : ")

maths = float(input("Enter maths marks : "))
science = float(input("Enter science marks : "))
english = float(input("Enter english marks : "))
sst = float(input("Enter sst marks : "))
hindi = float(input("Enter hindi marks : "))

total = maths + science + english + sst + hindi
percentage = (total / 500) * 100

if percentage >= 90:
    grade = "A"
elif percentage >= 80:
    grade = "B"
elif percentage >= 70:
    grade = "C"
elif percentage >= 60:
    grade = "D"
else:
    grade = "E"

print("Name : ", name)
print("Roll No : ", roll_no)
print("Total : ", total)
print("Percentage : ", percentage)
print("Grade : ", grade)
```
