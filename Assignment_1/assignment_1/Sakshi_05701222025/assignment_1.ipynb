```python
# find area of rectangle
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
area_rectangle = length * width
print("The area of the rectangle is: " , str(area_rectangle))
```

```python
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest: "))
time = float(input("Enter the time in years: "))
simple_interest = (principal * rate * time) / 100
print("The simple interest is: " , str(simple_interest))
```

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit: " , str(fahrenheit))
```

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
average = (num1 + num2 + num3) / 3
print("The average of the three numbers is: " , str(average))
```

```python
num = float(input("Enter a number: "))
square = num ** 2
cube = num ** 3
print("The square of the number is: " , str(square))
print("The cube of the number is: " , str(cube))
```

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
print("Before swapping: num1 = " + str(num1) + ", num2 = " + str(num2))
num1 = num1 + num2
num2 = num1 - num2
num1 = num1 - num2
print("After swapping: num1 = " + str(num1) + ", num2 = " + str(num2))
```

```python
name = input("Enter student's name: ")
age = int(input("Enter student's age: "))
marks_subject1 = float(input("Enter marks for subject 1: "))
marks_subject2 = float(input("Enter marks for subject 2: "))
marks_subject3 = float(input("Enter marks for subject 3: "))
total_marks = marks_subject1 + marks_subject2 + marks_subject3
percentage = (total_marks / 300) * 100
print("\nStudent Report")
print("Name: " + name)
print("Age: " + str(age))
print("Marks for Subject 1: " + str(marks_subject1))
print("Marks for Subject 2: " + str(marks_subject2))
print("Marks for Subject 3: " + str(marks_subject3))
print("Total Marks: " + str(total_marks))
print("Percentage: " + str(percentage) + "%")
```
