```python
#To find the area of rectangle
a=float(input("Enter length"))
b=float(input("Enter breadth"))
Area=a*b
print("The area of rectangle is :",Area)
```

```python
#find simple interest
P=float(input("Enter principle amount"))
R=float(input("Enter rate of interest"))
T=float(input("Enter time (e.g., 2 year or 2.5 year)"))
SI=(P*R*T)/100
print("The simple interest is :",SI)
```

```python
#To convert temperature from Celsius into Farenheit
T=float(input("Enter temperature in Celsius"))
F=(T*9/5)+32
print("The temperature in Farenheit is :",F)
```

```python
#To calculate average of 3 numbers
a=float(input("Enter first number"))
b=float(input("Enter second number"))
c=float(input("Enter third number"))
average=(a+b+c)/3
print("The average of 3 numbers is :",average)
```

```python
#To find square and cube of a number
a=float(input("Enter a number"))
square=a**2
cube=a**3
print("The square and cube of the number respectively are :",square,cube)
```

```python
#To swap two numbers without third variable
a=int(input("Enter a number:"))
b=int(input("Enter a number:"))
a,b=b,a
print("After swapping :a=",a,"b= ",b)
```

```python
#To create a student report program
# Student Report Program

name = input("Enter student name: ")
roll_no = input("Enter roll number: ")

english = float(input("Enter English marks: "))
maths = float(input("Enter Maths marks: "))
science = float(input("Enter Science marks: "))
sanskrit = float(input("Enter Sanskrit marks: "))

total = english + maths + science + sanskrit
percentage = (total / 400) * 100

print("\n----- STUDENT REPORT -----")
print("Name:", name)
print("Roll Number:", roll_no)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
