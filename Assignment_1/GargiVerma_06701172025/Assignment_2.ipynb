```python
# Question 1: Sum of First 10 Natural Numbers

total_sum = 0

for i in range(1, 11):
    total_sum += i

print(total_sum)
```

```python
# Question 2: Factorial of a Number

num = 5
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial =", fact)
```

```python
# Question 3: Fibonacci Series

n = 10

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# Question 4: Largest Among Three Numbers

a = 25
b = 40
c = 18

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("Largest Number =", largest)
```

```python
# Question 7: Student Report Program

name = "Gargi Verma"

marks = [85, 90, 78, 88, 92]

total = marks[0] + marks[1] + marks[2] + marks[3] + marks[4]

percentage = total / len(marks)

print("Student Name:", name)
print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 90:
    grade = "A"
elif percentage >= 80:
    grade = "B"
elif percentage >= 70:
    grade = "C"
elif percentage >= 60:
    grade = "D"
else:
    grade = "F"

print("Grade:", grade)
```
