1Q. Sum of first 10 natural numbers:


```python
total = 0
for i in range(1, 11):
    total = total + i
print("Sum of ten natural numbers is: ", total)
```

Q2. Factorial of a number:


```python
n = int(input("Enter number: "))
fact = 1
for i in range(1, n+1):
    fact = fact * i
print(f"Factorial of the number {n} is: ", fact)
```

Q5. Student Result System


```python
name = input("Enter Name: ")
roll = input("Enter Roll No: ")

total = 0
for i in range(5):
    marks = int(input(f"Subject {i+1} marks: "))
    total = total + marks

percentage = (total / 500) * 100

if percentage >= 90:
    grade = "A"
elif percentage >= 75:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 40:
    grade = "D"
else:
    grade = "F"

print("Name:", name)
print("Roll No:", roll)
print("Total:", total)
print("Percentage:", percentage, "%")
print("Grade:", grade)
```
