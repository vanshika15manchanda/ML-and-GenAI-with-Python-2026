Q2. Create a function to calculate sum of first N natual numbers:

```python
def sum_of_N_natual_no(n):
    count = 0
    for i in range(n+1):
        count+=i
    print(f"the sum of first {n} natual numbers is: ", count)

n= int(input("Enter a number: "))
sum_of_N_natual_no(n)
```

Q3. Create a function to reverse a number: 

```python
def reverse_number(n):
    if n < 0:
        sign = -1
    else:
        sign = 1

    num = int(str(abs(n))[::-1])
    print(f"the reversed of {n} is : ",sign*num)

reverse_number(163)
reverse_number(-3672)
```

Q4. create a fuction to count digits in a number: 

```python
def count_digits(x):
    num_of_digits = len(str(x))
    print(f"The number of digits in {x} is:", num_of_digits)

count_digits(462862)
count_digits(657)
```

Q5. Create a function to check palindrome number:

```python
def check_palindrome_number(num):
    if str(num) == str(num)[::-1]:
        print(num, "is a palindrome number.")
    else:
        print(num, "is not a palindrome number.")

num = int(input("Enter a number: "))
check_palindrome_number(num)
check_palindrome_number(1212)
```

Q6. Create a function to generate a fibonacci series:

Q7. Create a Calculator using functions:

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    if b==0:
        return "Zero divison error, Check again!"
    else:
        return a/b
def exp(a,b):
    return a**b

print("\n------------- Calculator -------------")
print("1. Addition.")
print("2. Substraction.")
print("3. Multiplication.")
print("4. Division")
print("5. Power.")

choice = int(input("Take your choice from [1-5] with respect to which task you want to perform:"))
a = float(input("ENTER THE FIRST NUMBER: "))
b = float(input("ENTER THE SECOND NUMBER: "))

if choice == 1:
    print("The addition is: ", add(a,b))
elif choice == 2:
    print("The substraction is: ", sub(a,b))
elif choice == 3:
    print("The multiplication is: ", mul(a,b))
elif choice == 4:
    print("The division is: ", div(a,b))
elif choice == 5:
    print("The Power is: ", exp(a,b))
else:
    print("Invalid Choice!!")
```
