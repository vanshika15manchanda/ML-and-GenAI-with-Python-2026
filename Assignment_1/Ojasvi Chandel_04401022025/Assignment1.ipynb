Q7. Create a student report program that take student details using input(), store marks in variables, calculate total percentage, use comments, use proper indentation:  

```python
# STUDENT REPORT PROGRAM

name = input("ENTER THE STUDENT NAME: ")
rollno = int(input("ENTER THE ROLL NUMBER OF THE SUDENT: "))

# MARKS OF STUDENT PER SUBJECT
total = 0
for i in range(5):
    marks = float(input("ENTER THE MARKS OF EACH SUBJECT: "))
    total += marks

# PERCENTAGE OF THE STUDENT
per = (total/500)*100

# Student Report Display:
print("---------------------------------STUDENT REPORT---------------------------------")
print("Student Name: ", name)
print("Roll number: ", rollno)
print("total marks: ", total)
print("percentage: ", per)
```
