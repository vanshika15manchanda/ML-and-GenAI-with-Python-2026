```python
#1 Find area of a rectangle
length= int(input("Enter the length:")) #it takes length of the rectangle
breadth= int(input("Enter the breadth:")) #it takes breadth of rectangle
area= length*breadth #area is calculated
print("area of rectangle is:",area) #area is given
```

```python
#2 Find simple interest
P= int(input("enter principle amount:")) #principle amount is entered
R= int(input("enter rate:")) #rate is taken as input
T= int(input("enter time:")) #time is entered by user
SI= ((P*R*T)/100)
print("simple interest is:",SI)
```

```python
# Convert temperature of celcius to fahrenheit
t= int(input("enter temperature:")) #takes temperature in celcius as input
f=t*(9/5)+32 #celcius is converted into fahrenheit
print("temperature in fahrenheit is:",f) #give temperature in fahrenheit as output
```

```python
#4 Calculate average of three number
a= int(input("Enter first number:")) #takes first number as input
q= int(input("enter second number:")) #takes second number as input
d= int(input("enter third number:")) #takes third number as input
average= (a+q+d)/3
print("average is:",average) #gives the average of numbers as output
```

```python
#5 find square and cube of a number
num= int(input("enter a number")) # a number is taken as input
print("square of",num, "is",num**2,"\n","cube of",num,"is",num**3)
```

```python
#6 swap two numbers without third variable
n1= int(input("enter first number:")) #takes first number as input
n2= int(input("enter second number:")) #takes second number as input
print("before swapping first number is:",n1,"second number is:",n2)
n1,n2= n2,n1
print("after swapping first number is:",n1,"second number is:",n2)
```

```python
#7 create a student report program that takes students details using input(), store marks in variables,calculate total and percentage
detail= input("enter student marks in dictionary:") #takes marks of all subjects as input
marks_dict = eval(detail) #converts details into dictionary
marks= marks_dict.values() #get marks list
total= sum(marks) #calculates sum total of numbers
print("Total marks;",total,"\nPercentage",total/(len(marks))) #final result
```
