```python
#1. Find area of rectangle
l=int(input("Enter the length of rectangle: "))
b=int(input("Enter the breadth of rectangle: "))
area=l*b
print("Area of rectangle is:", area)
```

```python
#2. Find simple interest
p=int(input("Enter the principal amount: "))
r=int(input("Enter the rate of interest: "))
t=int(input("Enter the time period: "))
SI=(p*r*t)/100
print("Simple interest is:", SI)
```

```python
#3. Convert temperature from Celsius to Fahrenheit
C= int(input("Enter the temperature in Celsius: "))
F= (C*9/5)+32
print("Temperature in Fahrenheit is:", F)
```

```python
#4. Average of 3 numbers
a = int(input("Enter the value of first number: "))
b=int(input("Enter the value of second number: "))
c=int(input("Enter the value of third number: "))
avg=(a+b+c)/3
print("Average is: ", avg)
```

```python
#6. Swap 2 numbers without third variable
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
print("The numbers before swapping are:", a,"and",b)
a=a+b
b=a-b
a=a-b
print("The numbers after swapping are:", a,"and",b)
```
