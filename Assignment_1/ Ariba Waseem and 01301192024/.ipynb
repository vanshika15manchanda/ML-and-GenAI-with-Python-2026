```python
# 1. Find area of rectangle.

# Naming the variables (identifier) and taking the inputs from user
length = float(input("Enter the length of rectangle: "))
breadth = float(input("Enter the breadth of rectangle: "))
# Applying the formula of area
Area = length * breadth
# Printing the area (output)
print("Area of rectangle is: ", Area)
```

```python
# 2. Find simple interest

# Naming the variables (identifier) and taking the inputs from user
Principle = float(input("Enter the interest: "))
Rate_of_Interest = float(input("Enter the rate of interest: "))  # using underscores in variable's name
Time = float(input("Enter the time: "))
# Applying the formula of simple interest
Simple_Interest = (Principle * Rate_of_Interest * Time) / 100
# Printing the simple interest
print("Simple Interest is: ", Simple_Interest)
```

```python
# 3. Convert temperature from celsius to farenheit

# Naming the variables (identifier) and taking the inputs from user
temp_in_cel = float(input("Enter the temperature in celsius: "))
# Applying the formula of celsius to farenheit
temp_in_far = (temp_in_cel * 9/5) + 32
# Printing the temperature
print('The temperature in farenheit is: ', temp_in_far)
```

```python
# 4. Calculate average of three numbers.

# Naming the variables (identifier) and taking the inputs from user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
# Applying the formula of average
average = (num1 + num2 + num3) / 3
# Printing the average
print("The average of three numbers is: ", average)
```

```python
# 5. Find square and cube of the numbers.

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
print("The square of the first number is: ", num1**2)
print("The square of the second number is: ", num2**2)
print("The cube of the first number is: ", num1**3)
print("The cube of the second number is: ", num2**3)
```

```python
# 6. Swap two numbers without using third variable.

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
a = a + b
b = a - b
a = a - b
print("The first number after swapping is: ", a)
print("The second number after swapping is: ", b)
```

```python
# 7. Create a student report program that takes student details using input(). Store marks in variables. Calculate the total and percentage. Use comments. Use proper indentations.

# Naming the variables (identifier) and taking the inputs from user
marks1 = float(input("Enter the marks of first subject: "))
marks2 = float(input("Enter the marks of second subject: "))
marks3 = float(input("Enter the marks of third subject: "))
marks4 = float(input("Enter the marks of fourth subject: "))
marks5 = float(input("Enter the marks of fifth subject: "))

# Calculating the total marks out of 500
Total = marks1 + marks2 + marks3 + marks4 + marks5
print("The total marks out of 500 are: ", Total)

# Calculating the percentage
Percentage = (Total / 500) * 100
print("The percentage is: ", Percentage)

# Commenting on report using proper indentation
if Percentage >= 90:
    print("Grade: A")
elif Percentage >= 80:
    print("Grade: B")
elif Percentage >= 70:
    print("Grade: C")
elif Percentage >= 60 and Percentage <= 50:
    print("Grade: D")
else:
    print("Fail")
```
